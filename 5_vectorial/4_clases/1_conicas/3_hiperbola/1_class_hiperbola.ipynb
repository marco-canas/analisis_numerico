````python
import pandas as pd
import numpy as np

# Cargar la lista de estudiantes desde el archivo CSV
path = 'C:/Users/marco/Documentos/docencia/groups_list/vectorial.xlsx'
estudiantes = pd.read_excel(path)
estudiantes.head()
````

````python

# Convertir la columna 'Nombre' a una lista de Python
lista_estudiantes = estudiantes['Nombre'].tolist()

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def asistencia():
    global lista_estudiantes
    if lista_estudiantes:
        estudiante_aleatorio = np.random.choice(lista_estudiantes)
        print(f"Llamando a {estudiante_aleatorio}...")
        asistencia = input(f"¿Asistió {estudiante_aleatorio}? (s/n): ").strip().lower()
        
        # Validación de entrada
        if asistencia == 's':
            # Obtener el índice del estudiante en el DataFrame
            indice = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index[0]
            # Incrementar su valor en la columna 'Parcial'
            estudiantes.at[indice, 'Parcial'] += 0.3
            print(f"Actualizado: {estudiante_aleatorio} ahora tiene {estudiantes.at[indice, 'Parcial']} en la columna 'Parcial'.")
        
        elif asistencia != 'n':
            print("Entrada inválida. Por favor, introduce 's' para sí o 'n' para no.")
            return  # Volver a preguntar para el mismo estudiante si la entrada es inválida
        
        # Remover al estudiante de la lista para no llamarlo de nuevo
        lista_estudiantes.remove(estudiante_aleatorio)
    else:
        print("Se han llamado a todos los estudiantes del grupo completo.")

# Función para guardar cambios al final de la sesión
def guardar_cambios():
    estudiantes.to_csv(path, index=False)
    print("Cambios guardados en el archivo CSV.")

# Ejemplo de uso
while lista_estudiantes:
    asistencia()

# Guardar los cambios una vez que todos los estudiantes hayan sido llamados
guardar_cambios()

````

### **Metodología de Enseñanza: Aprendizaje Basado en Problemas y Proyectos (ABPP)**



El **Aprendizaje Basado en Problemas y Proyectos (ABPP)** es una metodología educativa que integra los enfoques del **Aprendizaje Basado en Problemas (ABP)** y el **Aprendizaje Basado en Proyectos (PBL, por sus siglas en inglés)**. 

Es especialmente útil para la formación de futuros docentes, ingenieros y administradores, ya que combina el desarrollo de habilidades prácticas con el pensamiento crítico y reflexivo. A continuación, se presenta una descripción adaptada a la enseñanza de las matemáticas en educación infantil, Licenciatura en Matemáticas, Ingeniería Agropecuaria y Administración de Empresas:

---



### **1. Fundamentos del ABPP**


El ABPP se basa en:
- **Resolución de Problemas Reales:** Los estudiantes analizan y resuelven problemas auténticos relacionados con su área de formación, fomentando la aplicación práctica del conocimiento.
- **Proyectos como Herramientas Integradoras:** Los proyectos permiten desarrollar habilidades interdisciplinarias y ofrecer soluciones concretas a los problemas planteados.


- **Aprendizaje Activo:** Los estudiantes son responsables de su aprendizaje y participan activamente en el proceso de investigación, diseño y solución.

---

### **Características de funciones hiperbólicas en un modelo de riego**  
Las funciones hiperbólicas (como $$ y = \frac{a}{x}$$ o $$x^2 - y^2 = c^2$$) generan curvas que suelen aparecer en la naturaleza cuando hay dispersión desde un punto central o distribución en forma de lóbulos. 

En un modelo de riego, podríamos esperar:  
- **Distribución del agua en forma de hipérbolas**: Canales o trayectorias de riego que se expanden siguiendo estas curvas.  
- **Simetría hiperbólica**: Zonas donde el agua se propaga de forma decreciente, creando líneas similares a asíntotas.

### **Navegación GPS en Tractores y su Relación con las Hipérbolas**  

La navegación GPS en tractores es una tecnología clave en la **agricultura de precisión**, permitiendo a los agricultores optimizar el uso de recursos como fertilizantes, agua y combustible al mejorar la precisión en la siembra, la cosecha y otras operaciones en el campo.  

#### **¿Cómo funciona el GPS en tractores?**  
El sistema de posicionamiento global (**GPS, por sus siglas en inglés**) determina la ubicación de un tractor mediante señales enviadas desde **múltiples satélites en órbita**. Cada satélite transmite su posición y la hora exacta en que la señal fue enviada.  

El receptor GPS en el tractor mide el tiempo que tarda cada señal en llegar, lo que permite calcular la distancia a cada satélite. Para determinar con precisión la ubicación del tractor en el campo, el GPS usa un principio basado en la **trilateración**.  

---



### **Relación con las Hipérbolas**  
El principio matemático detrás de la navegación GPS está directamente relacionado con las **hipérbolas**, una de las secciones cónicas estudiadas en geometría analítica.  

1. **Tiempo de llegada de las señales**  
   - El GPS mide la **diferencia de tiempo** entre la llegada de las señales desde al menos **tres satélites**.  
   - La diferencia de tiempo entre dos señales define un **lugar geométrico** de posibles ubicaciones del receptor: **una hipérbola**.  

2. **Hipérbolas en la Geolocalización**  
   - Si el receptor recibe señales de **dos satélites**, se puede determinar que su ubicación está en **alguna parte de la hipérbola** formada por las diferencias de distancia a estos satélites.  
   - Con una tercera señal (de un tercer satélite), la intersección de las hipérbolas permite determinar una única posición en el espacio.  

3. **Analogía con la Definición de Hipérbola**  
   - Matemáticamente, una hipérbola es el conjunto de puntos cuya **diferencia de distancias** a dos focos es constante.  
   - En el GPS, los **focos** son los satélites, y la diferencia de tiempos de llegada de las señales se traduce en la diferencia de distancias, lo que forma hipérbolas que ayudan a localizar el tractor con precisión.  

---



### **Beneficios del GPS en la Agricultura de Precisión**  
- **Optimización del uso de insumos** (fertilizantes, herbicidas, semillas).  
- **Mejor planificación y eficiencia** en la siembra y cosecha.  
- **Reducción de costos** mediante trayectorias optimizadas.  
- **Mayor precisión** en el guiado automático del tractor, mejorando la calidad del cultivo.  

En resumen, el GPS en la agricultura usa el concepto de hipérbolas para calcular ubicaciones con alta precisión, facilitando la automatización y la toma de decisiones basadas en datos en el sector agropecuario. 🚜📡🌾

### **Justificación de la forma hiperbólica**  
1. **Optimización del flujo de aire**  
   - La geometría hiperbólica crea un **efecto Venturi**, acelerando el aire en la parte estrecha del silo y favoreciendo la circulación natural del aire.  
   - Esto permite que el aire caliente suba y el aire fresco entre por las aberturas inferiores.  

2. **Mayor ventilación y enfriamiento**  
   - La estructura hiperbólica genera **corrientes convectivas**, donde el aire fresco es atrapado en la parte inferior y asciende, expulsando el aire caliente.  
   - Este diseño ayuda a **reducir la temperatura interna**, manteniendo mejores condiciones para almacenar granos o productos agrícolas.  

3. **Resistencia estructural y eficiencia de materiales**  
   - Las curvas hiperbólicas ofrecen una **mayor resistencia estructural**, similar a la de los enfriadores de torres de refrigeración.  
   - Se usa menos material en comparación con una estructura cilíndrica con la misma capacidad.  

### **Conclusión**  
El diseño hiperbólico del silo no solo es estéticamente llamativo, sino que también **mejora la ventilación natural**, **mantiene el contenido fresco** y **reduce costos estructurales**. Es una solución eficiente para el almacenamiento agrícola en climas cálidos.

## Referentes 

La **teoría de representación semiótica** de Raymond Duval es fundamental en la didáctica de las matemáticas, especialmente en el estudio de los procesos cognitivos relacionados con el aprendizaje de conceptos matemáticos. Aunque el libro *"Semiosis y pensamiento humano"* (disponible en el enlace que compartes) es una fuente principal, existen referentes de acceso gratuito que pueden ayudarte a comprender su teoría.  

### **Referentes gratuitos sobre la teoría de Duval:**  

#### **1. Artículos y capítulos de acceso abierto:**  
- **Duval, R. (1993).** *"Registros de representación semiótica y funcionamiento cognitivo del pensamiento"*.  
  - Disponible en repositorios académicos como [ResearchGate](https://www.researchgate.net/) o [Academia.edu](https://www.academia.edu/).  
- **Duval, R. (1999).** *"Semiosis y pensamiento humano: Registros semióticos y aprendizajes intelectuales"*.  
  - Algunos extractos pueden encontrarse en [Google Scholar](https://scholar.google.com/).[SEMIOSISY PENSAMIENTO HUMANO. Registros semióticos y Aprendizajes intelectuales Raymond Duval](https://toaz.info/doc-view-3)  

#### **2. Tesis y documentos académicos:**  
- Varias universidades publican tesis que analizan la teoría de Duval. Por ejemplo:  
  - **Repositorio UNAL (Colombia):** [https://repositorio.unal.edu.co/](https://repositorio.unal.edu.co/) (buscar "Duval representación semiótica").  
  - **Repositorio UNAM (México):** [https://www.repositorio.unam.mx/](https://www.repositorio.unam.mx/).  

#### **3. Presentaciones y conferencias en línea:**  
- Algunas ponencias sobre Duval están disponibles en **YouTube** o **SlideShare**.  
  - Ejemplo: *"Registros semióticos en la enseñanza de las matemáticas"* (buscar en plataformas educativas).  

#### **4. Libros relacionados en acceso abierto:**  
- **"Visualización en matemáticas: Un enfoque desde la teoría de Duval"** (algunos capítulos pueden estar en [LibreTexts](https://es.libretexts.org/) o [OpenEdition](https://www.openedition.org/)).  

### **Alternativas si no puedes acceder al libro de pago:**  
- Revisa si tu universidad o biblioteca local tiene convenio con la **Editorial Univalle**.  
- Busca en **Library Genesis** ([https://libgen.is/](https://libgen.is/)) o **Z-Library** (dominios cambian frecuentemente).
