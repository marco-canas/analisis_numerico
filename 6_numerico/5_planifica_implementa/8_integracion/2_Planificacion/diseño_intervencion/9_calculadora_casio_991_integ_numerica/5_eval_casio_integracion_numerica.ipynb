```python
def riemann_precision(f, a, b, epsilon=0.0001):
    n = 1
    while True:
        dx = (b - a) / n
        sum_prev = sum(f(a + i * dx) * dx for i in range(n))
        sum_next = sum(f(a + i * dx) * dx for i in range(n+1)))
        if abs(sum_next - sum_prev) < epsilon:
            return n
        n += 1
```
