## **5. Contenido (Problema Real → Modelo Matemático)**  




**Problema:** *"Un ingeniero agropecuario necesita calcular el cero de la $c(x) = x^{3} - 2x - 5$ $kg/ha$ para optimizar el riego en un cultivo de arroz."*  


# Contemos de donde puede venir este problema  

### **Deducción del Modelo Matemático para Optimización de Riego en Cultivos de Arroz**  


**Contexto real en Antioquia, Colombia**:  
En la región de **El Bajo Cauca antioqueño**, los ingenieros agropecuarios enfrentan el desafío de determinar el **volumen óptimo de agua** (en m³/ha) para cultivos de arroz, donde el exceso o defecto de riego afecta la productividad.  





#### **1. Situación Real**  
Supongamos que:  
- **La tasa de crecimiento del arroz** (en kg/ha) sigue una relación cúbica con el volumen de agua aplicada ($x$).  
- **Datos empíricos** muestran que:  
  - Con $x = 1$ m³/ha → Crecimiento = $-6$ kg/ha (déficit hídrico).  
  - Con $x = 2$ m³/ha → Crecimiento = \(-1\) kg/ha (aún insuficiente).  
  - Con $x = 3$ m³/ha → Crecimiento = \(16\) kg/ha (óptimo).  



**Problema**:  
Encontrar el volumen exacto $x$ donde el crecimiento es **cero** (punto de transición entre déficit y crecimiento positivo), para ajustar el riego sin desperdiciar agua.  





#### **2. Construcción del Modelo Matemático**  
Se propone una función cúbica que modele el crecimiento:  
$$
f(x) = x^3 - 2x - 5
$$  



**¿Por qué esta función?**  
- **Condiciones dadas**:  
  1. $f(1) = 1^3 - 2(1) - 5 = -6$ ✔️  
  2. $f(2) = 8 - 4 - 5 = -1$ ✔️  
  3. $f(3) = 27 - 6 - 5 = 16$ ✔️  



- **Interpretación**:  
  - La raíz de $f(x) = 0$ ($x \approx 2.0946$) indica el **volumen crítico** donde el crecimiento pasa de negativo a positivo.  





#### **3. Deducción Gráfica**  


````python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 3, 100)
f = lambda x: x**3 - 2*x - 5

plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label='$f(x) = x^3 - 2x - 5$', color='blue')
plt.axhline(0, color='black', linestyle='--')
plt.scatter([2.0946], [0], color='red', label='Raíz ($x \\approx 2.0946$)')
plt.xlabel('Volumen de agua (m³/ha)')
plt.ylabel('Crecimiento del arroz (kg/ha)')
plt.title('Modelo de Crecimiento vs. Riego en Arroz')
plt.legend()
plt.grid(True)
plt.show()
````

 
**Salida**:  
- La gráfica muestra que la raíz está cerca de \(x = 2.0946\), el punto exacto donde el crecimiento es cero.  





#### **4. Relación con el Método Newton-Raphson**  
**Aplicación**:  
Para encontrar numéricamente la raíz:  
1. **Derivada**: $f'(x) = 3x^2 - 2$.  
2. **Iteración**:  
   $$
   x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
   $$  
   - Con $x_0 = 2$:  
     - $x_1 = 2 - \frac{-1}{10} = 2.1$  
     - $x_2 = 2.1 - \frac{0.061}{11.23} \approx 2.0946$  



**Resultado**:  
El volumen óptimo es **2.0946 m³/ha**, donde el cultivo comienza a tener crecimiento positivo.

### **Construcción del Modelo:**  


1. **Representación Verbal:** "Queremos encontrar $x$ tal que $f(x) = 0$."  
     1. Se traza un esbozo del gráfico de $f$ en el un intervalo $[a,b]$ donde se estime que se encuentra un cero de $f$. 
     2. A partir de la lectura al esbozo del gráfico de $f$, se estima una primera aproximación $x_{0}$ al cero de $f$ que buscamos. 
     3. Trazo el punto $(x_{0},f(x_{0}))$ en el gráfico de $f$.
     4. Determino la pendiente de la recta tangente a $f$ en el punto $x_{0}$.
   
     5. Determino la función $g(x) = f(x_{0}) + f'(x_{0})(x - x_{0})$ que corresponde a la recta tangente al grafico de $f$ en $(x_{0}, f(x_{0}))$.
     6. Determino el cero $x_{1}$ de la función lineal $g$ anterior, el cual será la segunda aproximación al cero de $f$ que buscamos. 
     7. Repito los pasos 3, 4, 5 y 6 hasta que la diferencia entre i-ésima aproximación $x_{i}$ y la anterior $x_{i-1}$ sea menor que un valor dado $\epsilon$ que se estima como la precisión deseada.

Prompt: Dame, para la función $f(x) = e^{x} - x$,, la secuencia gráfica en matplotlib, donde en  casa subplot este representada una de tres iteraciones del método de newton - raphson para la determinación del cero de $f$

````python
import numpy as np
from scipy.optimize import fsolve

import matplotlib.pyplot as plt

# Definir la función
f = lambda x: np.exp(x) + x

# Crear valores de x
a,b = -1,0
x_vals = np.linspace(a, b, 500)

# Calcular valores de f(x)
y_vals = f(x_vals)

# Encontrar el cero de la función
x_zero = fsolve(f, -0.5)  # Aproximación inicial cerca del cero

# Graficar la función
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='$f(x) = e^x + x$', color='blue')
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)  # Línea horizontal en y=0
plt.scatter(x_zero, [0], color='red', label=f'Cero: $x = {x_zero[0]:.4f}$', zorder=5)  # Resaltar el cero

# Configuración del gráfico
plt.title('Gráfico de $f(x) = e^x + x$', fontsize=14)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$f(x)$', fontsize=12)
plt.legend(fontsize=12)
plt.grid()
plt.show()
````

# Código adaptado para la función $f(x) = e^x + x$ con visualización de las tres primeras iteraciones


````python
# Seguencia gráfica del algoritmo de Newton para aproximar ceros de funciones 

import numpy as np
import matplotlib.pyplot as plt

# Definimos la nueva función y su derivada
f = lambda x: np.exp(x) + x  # Cambiamos la función aquí
df = lambda x: np.exp(x) + 1  # Derivada actualizada

# Punto inicial
x0 = -1.0  # Punto inicial modificado para mejor visualización
iteraciones = 3  # Mostraremos solo 3 iteraciones
a,b = -1.1, -0.4
# Configuración de la figura
fig, axes = plt.subplots(1, iteraciones, figsize=(18, 5))
x_vals = np.linspace(a, b, 100)  # Rango ajustado para mejor visualización

# Iteraciones de Newton-Raphson
for i in range(iteraciones):
    x1 = x0 - f(x0) / df(x0)  # Fórmula de Newton-Raphson
    
    # Gráfica de la función
    axes[i].plot(x_vals, f(x_vals), label='$f(x) = e^x + x$', color='blue', linewidth=2)
    axes[i].axhline(y = 0, color='black', linewidth=0.8, linestyle='--') # eje x 
    
    # Tangente en x0
    tangent = f(x0) + df(x0) * (x_vals - x0) # ecuación de la recta tangente
    axes[i].plot(x_vals, tangent, label=f'Tangente en $x_{i}$', color='red', linestyle='--', linewidth=1.5) # grafica de la recta tangente
    
    # Puntos de interés
    axes[i].scatter([x0], [0], color='green', s=100, label=f'$x_{i}$', zorder=5) # primera aproximación al cero 
    axes[i].scatter([x0], [f(x0)], color='red', s=100, label=f'$(x_{i}, f(x_{i}))$', zorder=5) # punto de tangencia
    axes[i].scatter([x1], [0], color='purple', s=100, label=f'$x_{i+1}$', zorder=5) # la siguiente aproximación al cero 
    
    # Línea vertical para mostrar la aproximación
    axes[i].axvline(x1, color='gray', linestyle=':', alpha=0.5)
    
    # Configuración del subplot
    axes[i].set_title(f'Iteración {i+1}\n$x_{i} = {x0:.4f}$ → $x_{i+1} = {x1:.4f}$', pad=15)
    axes[i].legend(loc='lower right')
    axes[i].grid(True, alpha=0.3)
    axes[i].set_xlabel('x')
    if i == 0:
        axes[i].set_ylabel('f(x)')
    
    # Actualizar x0 para la siguiente iteración
    x0 = x1

plt.suptitle('Método de Newton-Raphson para $f(x) = e^x + x$', y=1.05, fontsize=14)
plt.tight_layout()
save_fig('secuencia_grafica_met_newton')
plt.show()

````

````python
import numpy as np
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

def newton_raphson_analysis(f, df, x0):
    """
    Realiza un análisis del método de Newton-Raphson para las primeras tres iteraciones.
    
    Parámetros:
        f: Función lambda para la cual se busca la raíz.
        df: Derivada de la función lambda.
        x0: Primera aproximación inicial.
    
    Retorna:
        - Secuencia de 4 gráficos mostrando las iteraciones.
        - Descripción verbal de las aproximaciones.
        - Representación tabular del proceso iterativo.
        - Script de Python asociado.
    """
    # Inicialización
    iteraciones = 3
    historia = []
    r = 1 # radio del intervalo centrado en x0
    x_vals = np.linspace(x0 - r, x0 + r, 500)
    
    # Gráficos
    fig, axes = plt.subplots(1, iteraciones, figsize=(18, 5))
    
    for i in range(iteraciones):
        x1 = x0 - f(x0) / df(x0)  # Fórmula de Newton-Raphson
        historia.append((i + 1, x0, f(x0), x1))
        
        # Gráfica de la función
        axes[i].plot(x_vals, f(x_vals), label='$f(x)$', color='blue')
        axes[i].axhline(0, color='black', linestyle='--', linewidth=0.8)
        
        # Tangente en x0
        tangent = f(x0) + df(x0) * (x_vals - x0)
        axes[i].plot(x_vals, tangent, label=f'Tangente en $x_{i}$', color='red', linestyle='--')
        
        # Puntos de interés
        axes[i].scatter([x0], [f(x0)], color='red', label=f'$(x_{i}, f(x_{i}))$')
        axes[i].scatter([x1], [0], color='green', label=f'$x_{i+1}$')
        
        # Configuración del subplot
        axes[i].set_title(f'Iteración {i + 1}')
        axes[i].legend()
        axes[i].grid(True, alpha=0.3)
        
        # Actualizar x0 para la siguiente iteración
        x0 = x1
    
    plt.suptitle('Método de Newton-Raphson: Secuencia de Iteraciones', fontsize=16)
    plt.tight_layout()
    plt.show()
    
    # Descripción verbal
    descripcion = "Descripción Verbal:\n"
    for i, (n, x_prev, fx_prev, x_next) in enumerate(historia):
        descripcion += f"Iteración {n}: Partiendo de x_{n-1} = {x_prev:.5f}, se calcula f(x_{n-1}) = {fx_prev:.5f}.\n"
        descripcion += f"Se traza la tangente en x_{n-1} y se encuentra la siguiente aproximación x_{n} = {x_next:.5f}.\n\n"
    print(descripcion)
    
    # Representación tabular
    tabla = pd.DataFrame(historia, columns=['Iteración', 'x_n', 'f(x_n)', 'x_n+1'])
    print("Representación Tabular:")
    print(tabla)
    
    # Script asociado
    script = """
import matplotlib.pyplot as plt

# Definir la función y su derivada
f = lambda x: ...
df = lambda x: ...

# Parámetros iniciales
x0 = ...
iteraciones = 3

# Gráficos
x_vals = np.linspace(x0 - 2, x0 + 2, 500)
for i in range(iteraciones):
    x1 = x0 - f(x0) / df(x0)
    tangent = f(x0) + df(x0) * (x_vals - x0)
    plt.plot(x_vals, f(x_vals), label='$f(x)$', color='blue')
    plt.plot(x_vals, tangent, label=f'Tangente en $x_{i}$', color='red', linestyle='--')
    plt.scatter([x0], [f(x0)], color='red', label=f'$(x_{i}, f(x_{i}))$')
    plt.scatter([x1], [0], color='green', label=f'$x_{i+1}$')
    plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
    plt.title(f'Iteración {i + 1}')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
    x0 = x1
"""
    print("\nScript de Python asociado:")
    print(script)
````

````python
newton_raphson_analysis(lambda x:np.exp(x) + x, lambda x: np.exp(x) + 1, 1)
````



## **6. Función Python para Iteraciones con Precisión**  


````python
def newton_raphson(f, df, x0, epsilon=0.0001, max_iter=100):
    """
    Encuentra la raíz de f(x) = 0 con precisión epsilon.
    
    Parámetros:
        f: Función a resolver.
        df: Derivada de f.
        x0: Punto inicial.
        epsilon: Precisión deseada.
        max_iter: Máximo de iteraciones.
    
    Retorna:
        (raíz, iteraciones, historia)
    """
    historia = []
    for n in range(max_iter):
        x1 = x0 - f(x0)/df(x0)
        historia.append((x0, x1))
        if abs(x1 - x0) < epsilon:
            return x1, n+1, historia
        x0 = x1
    raise ValueError("No converge en {} iteraciones".format(max_iter))
````

````python

# Ejemplo de uso:
f = lambda x: x**3 - 2*x - 5
df = lambda x: 3*x**2 - 2
raiz, iteraciones, _ = newton_raphson(f, df, x0=2.0)
print(f"Raíz encontrada: {raiz:.5f} en {iteraciones} iteraciones")

````

### **Implementación en Python (con Visualización)**  

Esta visualización muestra a la colección de puntos $(x_{i}, f(x_{i}))$ del método de Newton.


````python
import matplotlib.pyplot as plt
import numpy as np

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import matplotlib

# Configuración específica para VS Code
matplotlib.rcParams['animation.html'] = 'jshtml'
%matplotlib inline

# Definimos la función y su derivada
f = lambda x: np.exp(x) + x  # Función ejemplo: e^x + x
df = lambda x: np.exp(x) + 1  # Su derivada

def animate_newton(f, df, x0, epsilon, max_iter=10):
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Rango de visualización adaptativo
    x_min, x_max = x0-0.5, x0+0.5
    x_vals = np.linspace(x_min, x_max, 400)
    
    # Graficar la función y el eje x
    ax.plot(x_vals, f(x_vals), label=f'f(x) = $e^x + x$', color='blue')
    ax.axhline(0, color='black', linestyle='--', linewidth=0.8)
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.set_title('Método de Newton-Raphson')
    ax.grid(True, alpha=0.3)
    ax.legend()
    
    # Inicializar variables para la animación
    points, = ax.plot([], [], 'ro', label='Aproximaciones')
    lines, = ax.plot([], [], 'r--', linewidth=1.5, alpha=0.7)
    
    # Lista para almacenar el historial de iteraciones
    history = []
    
    def init():
        points.set_data([], [])
        lines.set_data([], [])
        return points, lines
    
    def update(frame):
        nonlocal x0  # Permite modificar x0 en el scope exterior
        
        # Calcular nueva aproximación
        x1 = x0 - f(x0)/df(x0)
        
        # Almacenar el paso actual
        history.append((x0, x1))
        
        # Actualizar gráficos
        x_points = [x for x, _ in history]
        y_points = [f(x) for x, _ in history]
        points.set_data(x_points, y_points)
        
        # Dibujar líneas tangentes
        if len(history) > 0:
            tangent_x = np.linspace(x0-0.5, x0+0.5, 2)
            tangent_y = f(x0) + df(x0)*(tangent_x - x0)
            lines.set_data(tangent_x, tangent_y)
        
        # Actualizar x0 para la siguiente iteración
        x0 = x1
        
        # Ajustar límites del gráfico si es necesario
        ax.set_xlim(min(x_vals.min(), x1-1), max(x_vals.max(), x1+1))
        
        return points, lines
    
    # Crear la animación
    anim = FuncAnimation(fig, update, frames=range(max_iter),
                        init_func=init, blit=True, interval=800,
                        repeat=False)
    
    # Cerrar la figura para evitar duplicados en VS Code
    plt.close()
    
    return anim

# Ejemplo de uso
animation = animate_newton(f, df, x0=-1.0, epsilon=1e-6, max_iter=5)
HTML(animation.to_jshtml())
````

## **Proyecto 1: Aproximación de Raíces en Polinomios Agrícolas**  
**Contexto:** Modelar el crecimiento de cultivos con funciones polinómicas.  
**Ejemplo:**  Leonel
- Función: $f(x) = x^3 - 2x^2 - 5x + 6$ (raíces relacionadas con puntos críticos de crecimiento).  
**Tareas:**  
1. Graficar la función en Python (`matplotlib`).  
2. Aplicar el método de Newton con $x_0 = 1.5$ y tolerancia $10^{-5}$.  
3. Comparar con la solución analítica.  
**Software:** Python (SymPy para derivadas).  



````python
from math import gcd

def posibles_ceros_racionales(a, b, c, d):
    """
    Para el polinomio f(x) = a*x^3 + b*x^2 + c*x + d,
    retorna la lista de posibles ceros racionales según el teorema de los ceros racionales.
    """
    def divisores(n):
        n = abs(n)
        divs = set()
        for i in range(1, n+1):
            if n % i == 0:
                divs.add(i)
                divs.add(-i)
        return divs

    p_divs = divisores(d)  # divisores del término independiente
    q_divs = divisores(a)  # divisores del coeficiente líder

    posibles = set()
    for p in p_divs:
        for q in q_divs:
            if q == 0:
                continue
            g = gcd(p, q)
            num = p // g
            den = q // g
            posibles.add(num / den)
    return sorted(posibles)

# Ejemplo de uso para f(x) = x^3 - 2x^2 - 5x + 6
ceros = posibles_ceros_racionales(1, +2, -5, 6)
print(ceros)
````
