## **5. Contenido (Problema Real ‚Üí Modelo Matem√°tico)**  




**Problema:** *"Un ingeniero agropecuario necesita calcular la ra√≠z de la ecuaci√≥n $f(x) = x^3 - 2x - 5$ para optimizar el riego en un cultivo de arroz."*  


# Contemos de donde puede venir este problema  

### **Deducci√≥n del Modelo Matem√°tico para Optimizaci√≥n de Riego en Cultivos de Arroz**  
**Contexto real en Antioquia, Colombia**:  
En la regi√≥n de **El Bajo Cauca antioque√±o**, los ingenieros agropecuarios enfrentan el desaf√≠o de determinar el **volumen √≥ptimo de agua** (en m¬≥/ha) para cultivos de arroz, donde el exceso o defecto de riego afecta la productividad.  

---

#### **1. Situaci√≥n Real**  
Supongamos que:  
- **La tasa de crecimiento del arroz** (en kg/ha) sigue una relaci√≥n c√∫bica con el volumen de agua aplicada (\(x\)).  
- **Datos emp√≠ricos** muestran que:  
  - Con \(x = 1\) m¬≥/ha ‚Üí Crecimiento = \(-6\) kg/ha (d√©ficit h√≠drico).  
  - Con \(x = 2\) m¬≥/ha ‚Üí Crecimiento = \(-1\) kg/ha (a√∫n insuficiente).  
  - Con \(x = 3\) m¬≥/ha ‚Üí Crecimiento = \(16\) kg/ha (√≥ptimo).  

**Problema**:  
Encontrar el volumen exacto \(x\) donde el crecimiento es **cero** (punto de transici√≥n entre d√©ficit y crecimiento positivo), para ajustar el riego sin desperdiciar agua.  

---

#### **2. Construcci√≥n del Modelo Matem√°tico**  
Se propone una funci√≥n c√∫bica que modele el crecimiento:  
$$
f(x) = x^3 - 2x - 5
$$  

**¬øPor qu√© esta funci√≥n?**  
- **Condiciones dadas**:  
  1. \(f(1) = 1^3 - 2(1) - 5 = -6\) ‚úîÔ∏è  
  2. \(f(2) = 8 - 4 - 5 = -1\) ‚úîÔ∏è  
  3. \(f(3) = 27 - 6 - 5 = 16\) ‚úîÔ∏è  

- **Interpretaci√≥n**:  
  - La ra√≠z de \(f(x) = 0\) (\(x \approx 2.0946\)) indica el **volumen cr√≠tico** donde el crecimiento pasa de negativo a positivo.  

---

#### **3. Deducci√≥n Gr√°fica**  


````python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 3, 100)
f = lambda x: x**3 - 2*x - 5

plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label='$f(x) = x^3 - 2x - 5$', color='blue')
plt.axhline(0, color='black', linestyle='--')
plt.scatter([2.0946], [0], color='red', label='Ra√≠z ($x \\approx 2.0946$)')
plt.xlabel('Volumen de agua (m¬≥/ha)')
plt.ylabel('Crecimiento del arroz (kg/ha)')
plt.title('Modelo de Crecimiento vs. Riego en Arroz')
plt.legend()
plt.grid(True)
plt.show()
````

 
**Salida**:  
- La gr√°fica muestra que la ra√≠z est√° cerca de \(x = 2.0946\), el punto exacto donde el crecimiento es cero.  

---

#### **4. Relaci√≥n con el M√©todo Newton-Raphson**  
**Aplicaci√≥n**:  
Para encontrar num√©ricamente la ra√≠z:  
1. **Derivada**: \(f'(x) = 3x^2 - 2\).  
2. **Iteraci√≥n**:  
   $$
   x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
   $$  
   - Con $x_0 = 2$:  
     - $x_1 = 2 - \frac{-1}{10} = 2.1$  
     - $x_2 = 2.1 - \frac{0.061}{11.23} \approx 2.0946$  

**Resultado**:  
El volumen √≥ptimo es **2.0946 m¬≥/ha**, donde el cultivo comienza a tener crecimiento positivo.

- **Conexi√≥n real**: Evita el **sobrerriego** (costoso y da√±ino) y el **subriego** (p√©rdida de productividad).

### **Construcci√≥n del Modelo:**  


1. **Representaci√≥n Verbal:** "Queremos encontrar $x$ tal que $f(x) = 0$."  
     1. Se traza un esbozo del gr√°fico de $f$ en el un intervalo $[a,b]$ donde se estime que se encuentra un cero de $f$. 
     2. A partir de la lectura al esbozo del gr√°fico de $f$, se estima una primera aproximaci√≥n $x_{0}$ al cero de $f$ que buscamos. 
     3. Trazo el punto $(x_{0},f(x_{0}))$ en el gr√°fico de $f$.
     4. Determino la pendiente de la recta tangente a $f$ en el punto $x_{0}$.
   
     5. Determino la funci√≥n $g(x) = f(x_{0}) + f'(x_{0})(x - x_{0})$ que corresponde a la recta tangente al grafico de $f$ en $(x_{0}, f(x_{0}))$.
     6. Determino el cero $x_{1}$ de la funci√≥n lineal $g$ anterior, el cual ser√° la segunda aproximaci√≥n al cero de $f$ que buscamos. 
     7. Repito los pasos 3, 4, 5 y 6 hasta que la diferencia entre i-√©sima aproximaci√≥n $x_{i}$ y la anterior $x_{i-1}$ sea menor que un valor dado $\epsilon$ que se estima como la precisi√≥n deseada. 


2. **Representaci√≥n Geom√©trica:** Gr√°fica de $f(x)$ y aproximaci√≥n por tangentes.  

Prompt: Dame, para la funci√≥n $f(x) = e^{x} - x$,, la secuencia gr√°fica en matplotlib, donde en  casa subplot este representada una de tres iteraciones del m√©todo de newton - raphson para la determinaci√≥n del cero de $f$

````python
import numpy as np
from scipy.optimize import fsolve

import matplotlib.pyplot as plt

# Definir la funci√≥n
f = lambda x: np.exp(x) + x

# Crear valores de x
a,b = -1,0
x_vals = np.linspace(a, b, 500)

# Calcular valores de f(x)
y_vals = f(x_vals)

# Encontrar el cero de la funci√≥n
x_zero = fsolve(f, -0.5)  # Aproximaci√≥n inicial cerca del cero

# Graficar la funci√≥n
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='$f(x) = e^x + x$', color='blue')
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)  # L√≠nea horizontal en y=0
plt.scatter(x_zero, [0], color='red', label=f'Cero: $x = {x_zero[0]:.4f}$', zorder=5)  # Resaltar el cero

# Configuraci√≥n del gr√°fico
plt.title('Gr√°fico de $f(x) = e^x + x$', fontsize=14)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$f(x)$', fontsize=12)
plt.legend(fontsize=12)
plt.grid()
plt.show()
````

````python
# Seguencia gr√°fica del algoritmo de Newton para aproximar ceros de funciones 

import numpy as np
import matplotlib.pyplot as plt

# Definimos la nueva funci√≥n y su derivada
f = lambda x: np.exp(x) + x  # Cambiamos la funci√≥n aqu√≠
df = lambda x: np.exp(x) + 1  # Derivada actualizada

# Punto inicial
x0 = -1.0  # Punto inicial modificado para mejor visualizaci√≥n
iteraciones = 3  # Mostraremos solo 3 iteraciones
a,b = -1.1, -0.4
# Configuraci√≥n de la figura
fig, axes = plt.subplots(1, iteraciones, figsize=(18, 5))
x_vals = np.linspace(a, b, 100)  # Rango ajustado para mejor visualizaci√≥n

# Iteraciones de Newton-Raphson
for i in range(iteraciones):
    x1 = x0 - f(x0) / df(x0)  # F√≥rmula de Newton-Raphson
    
    # Gr√°fica de la funci√≥n
    axes[i].plot(x_vals, f(x_vals), label='$f(x) = e^x + x$', color='blue', linewidth=2)
    axes[i].axhline(y = 0, color='black', linewidth=0.8, linestyle='--') # eje x 
    
    # Tangente en x0
    tangent = f(x0) + df(x0) * (x_vals - x0) # ecuaci√≥n de la recta tangente
    axes[i].plot(x_vals, tangent, label=f'Tangente en $x_{i}$', color='red', linestyle='--', linewidth=1.5) # grafica de la recta tangente
    
    # Puntos de inter√©s
    axes[i].scatter([x0], [0], color='green', s=100, label=f'$x_{i}$', zorder=5) # primera aproximaci√≥n al cero 
    axes[i].scatter([x0], [f(x0)], color='red', s=100, label=f'$(x_{i}, f(x_{i}))$', zorder=5) # punto de tangencia
    axes[i].scatter([x1], [0], color='purple', s=100, label=f'$x_{i+1}$', zorder=5) # la siguiente aproximaci√≥n al cero 
    
    # L√≠nea vertical para mostrar la aproximaci√≥n
    axes[i].axvline(x1, color='gray', linestyle=':', alpha=0.5)
    
    # Configuraci√≥n del subplot
    axes[i].set_title(f'Iteraci√≥n {i+1}\n$x_{i} = {x0:.4f}$ ‚Üí $x_{i+1} = {x1:.4f}$', pad=15)
    axes[i].legend(loc='lower right')
    axes[i].grid(True, alpha=0.3)
    axes[i].set_xlabel('x')
    if i == 0:
        axes[i].set_ylabel('f(x)')
    
    # Actualizar x0 para la siguiente iteraci√≥n
    x0 = x1

plt.suptitle('M√©todo de Newton-Raphson para $f(x) = e^x + x$', y=1.05, fontsize=14)
plt.tight_layout()
save_fig('secuencia_grafica_met_newton')
plt.show()

````

## **6. Funci√≥n Python para Iteraciones con Precisi√≥n**  


````python
def newton_raphson(f, df, x0, epsilon=0.0001, max_iter=100):
    """
    Encuentra la ra√≠z de f(x) = 0 con precisi√≥n epsilon.
    
    Par√°metros:
        f: Funci√≥n a resolver.
        df: Derivada de f.
        x0: Punto inicial.
        epsilon: Precisi√≥n deseada.
        max_iter: M√°ximo de iteraciones.
    
    Retorna:
        (ra√≠z, iteraciones, historia)
    """
    historia = []
    for n in range(max_iter):
        x1 = x0 - f(x0)/df(x0)
        historia.append((x0, x1))
        if abs(x1 - x0) < epsilon:
            return x1, n+1, historia
        x0 = x1
    raise ValueError("No converge en {} iteraciones".format(max_iter))
````

````python

# Ejemplo de uso:
f = lambda x: x**3 - 2*x - 5
df = lambda x: 3*x**2 - 2
raiz, iteraciones, _ = newton_raphson(f, df, x0=2.0)
print(f"Ra√≠z encontrada: {raiz:.5f} en {iteraciones} iteraciones")

````

### **Implementaci√≥n en Python (con Visualizaci√≥n)**  

Esta visualizaci√≥n muestra a la colecci√≥n de puntos $(x_{i}, f(x_{i}))$ del m√©todo de Newton.


````python
import matplotlib.pyplot as plt
import numpy as np

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import matplotlib

# Configuraci√≥n espec√≠fica para VS Code
matplotlib.rcParams['animation.html'] = 'jshtml'
%matplotlib inline

# Definimos la funci√≥n y su derivada
f = lambda x: np.exp(x) + x  # Funci√≥n ejemplo: e^x + x
df = lambda x: np.exp(x) + 1  # Su derivada

def animate_newton(f, df, x0, epsilon, max_iter=10):
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Rango de visualizaci√≥n adaptativo
    x_min, x_max = x0-0.5, x0+0.5
    x_vals = np.linspace(x_min, x_max, 400)
    
    # Graficar la funci√≥n y el eje x
    ax.plot(x_vals, f(x_vals), label=f'f(x) = $e^x + x$', color='blue')
    ax.axhline(0, color='black', linestyle='--', linewidth=0.8)
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.set_title('M√©todo de Newton-Raphson')
    ax.grid(True, alpha=0.3)
    ax.legend()
    
    # Inicializar variables para la animaci√≥n
    points, = ax.plot([], [], 'ro', label='Aproximaciones')
    lines, = ax.plot([], [], 'r--', linewidth=1.5, alpha=0.7)
    
    # Lista para almacenar el historial de iteraciones
    history = []
    
    def init():
        points.set_data([], [])
        lines.set_data([], [])
        return points, lines
    
    def update(frame):
        nonlocal x0  # Permite modificar x0 en el scope exterior
        
        # Calcular nueva aproximaci√≥n
        x1 = x0 - f(x0)/df(x0)
        
        # Almacenar el paso actual
        history.append((x0, x1))
        
        # Actualizar gr√°ficos
        x_points = [x for x, _ in history]
        y_points = [f(x) for x, _ in history]
        points.set_data(x_points, y_points)
        
        # Dibujar l√≠neas tangentes
        if len(history) > 0:
            tangent_x = np.linspace(x0-0.5, x0+0.5, 2)
            tangent_y = f(x0) + df(x0)*(tangent_x - x0)
            lines.set_data(tangent_x, tangent_y)
        
        # Actualizar x0 para la siguiente iteraci√≥n
        x0 = x1
        
        # Ajustar l√≠mites del gr√°fico si es necesario
        ax.set_xlim(min(x_vals.min(), x1-1), max(x_vals.max(), x1+1))
        
        return points, lines
    
    # Crear la animaci√≥n
    anim = FuncAnimation(fig, update, frames=range(max_iter),
                        init_func=init, blit=True, interval=800,
                        repeat=False)
    
    # Cerrar la figura para evitar duplicados en VS Code
    plt.close()
    
    return anim

# Ejemplo de uso
animation = animate_newton(f, df, x0=-1.0, epsilon=1e-6, max_iter=5)
HTML(animation.to_jshtml())
````
