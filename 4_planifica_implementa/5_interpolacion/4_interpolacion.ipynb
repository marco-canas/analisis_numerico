
## 6. Funci√≥n para Suma de Riemann (Precisi√≥n Œµ)


````python
def riemann_n(f, a, b, epsilon=0.0001):
    """
    Determina el n√∫mero de rect√°ngulos n necesario para alcanzar la precisi√≥n epsilon
    en la integral aproximada de f entre a y b usando Sumas de Riemann.
    
    Args:
        f (function): Funci√≥n a integrar
        a, b (float): L√≠mites de integraci√≥n
        epsilon (float): Precisi√≥n deseada
    
    Returns:
        tuple: (n m√≠nimo, error estimado)
    """
    n = 1
    error = float('inf')
    integral_prev = 0
    
    while error > epsilon:
        n *= 2  # Estrategia de duplicaci√≥n
        x = np.linspace(a, b, n+1)
        delta = (b - a)/n
        integral = sum(f(a + i*delta) * delta for i in range(n))
        error = abs(integral - integral_prev)
        integral_prev = integral
        
    return n, error

# Ejemplo de uso:
def f(x): return x**2 + 2*x + 1
n, error = riemann_n(f, 0, 5)
print(f"Se requieren {n} rect√°ngulos para error < 0.0001")

````

El resultado de observaciones de un proceso de laboratorio o industrial genera una lista de datos que forman tablas.

Por ejemplo, la medici√≥n de la densidad del carbonato de sodio a distintas concentraciones a temperatura constante genera los siguientes datos.    


Los datos expresan la relaci√≥n de la densidad (variable dependiente) y la concentraci√≥n (variable independiente). 

Es decir que el valor de y depende del valor de x. Si deseamos conocer el valor de la variable dependiente para un valor de la variable independiente que no est√° en la tabla de datos, se requiere hacer una estimaci√≥n.

La estimaci√≥n puede lograrse de dos maneras, ajustando una funci√≥n que toque todos los puntos de la tabla (interpolaci√≥n) o ajustando una funci√≥n que minimice las distancias del punto tabulado a la funci√≥n (regresi√≥n).

La funci√≥n de la interpolaci√≥n pasa por cada punto de la tabla, la funci√≥n de la regresi√≥n no necesariamente pasa lo m√°s cerca de los puntos. 

Esto es, si sustituimos $x_{i}$ en $f(x)$, el resultado es $y_{i}$ para la interpolaci√≥n y para la regresi√≥n obtenemos $y_{i}$ corregida.

## 2.1 Interpolaci√≥n   

El objetivo es ajustar una funci√≥n que por combinaciones lineales logre pasar por todos los puntos.  

### Definici√≥n 2.1.1 ‚Äî Interpolaci√≥n. 

Las t√©cnicas de interpolaci√≥n ajustan un polinomio de menor grado que pasa por los puntos tabulados,esto quiere decir que, si sustituimos un valor de xi de la tabla, obtenemos la correspondiente yi de la tabla. En general, la funci√≥n tiene la forma:

$$ f(x) = a_{0}g_{0}(x) + a_{1}g_{1}(x) + a_{2}g_{2}(x) + a_{3}g_{3}(x) + \cdots + a_{n}g_{n}(x)  $$

Donde las $a_{i}$ son factores a determinar y las gi(x) son funciones de una familia en particular.

## Familia de monomios.   

Son los m√°s utilizados ya que son f√°ciles de obtener y generan un polinomio. La desventaja es que los polinomios tienen comportamientos diferentes a los esperados, de acuerdo con los datos, cuando son de potencias altas:  

$$ f(x) = a_{0} + a_{1}x + a_{2}x^{2} + a_{3}x^{3} + \cdots + a_{n}x^{n} $$

$$ f(x) = \sum_{i = 0}^{n} a_{i}x^{i} $$

La t√©cnica de ajustar un polinomio a los datos es la m√°s usada porque se adapta a muchos casos simples, las otras t√©cnicas se usan para un conjunto de datos que tienen un comportamiento c√≠clico o exponencial.

### 2.1.1 Interpolaci√≥n lineal  

La interpolaci√≥n lineal ajusta un polinomio de grado 1 a los datos tabulados. Usamos entonces la forma de una ecuaci√≥n lineal

$$ f(x) = a + bx $$

Para lograr una interpolaci√≥n lineal (polinomio de grado 1), se requieren dos puntos de la tabla  

|x|y|
|-|-|
|x_{0}|y_{0}|
|x_{1}|y_{1}|

Como indica la definici√≥n 2.1.1 se debe encontrar un polinomio que pase por los puntos tabulados. Entonces, si sustituimos x0 en f (x) (Ecuaci√≥n lineal 2.2 ) debemos obtener y0.

$$ f(x_{0}) = a + bx_{0} = y_{0} $$
Si sustituimos x1 en f (x) (Ecuaci√≥n lineal 2.2 ) debemos obtener y1.
$$ f(x_{1}) = a + bx_{1} = y_{1} $$

Tenemos dos ecuaciones (2.3 y 2.4) y dos inc√≥gnitas (a y b), entonces procedemos a resolver el sistema de 2x2.

Despejamos a de ecuaci√≥n 2.3 

$$ a = y_{0} - bx_{0} $$

Sustituimos a en la ecuaci√≥n 2.4
$$ y_{1} = (y_{0} - bx_{0}) + bx_{1} $$

$$ b = \frac{y_{1}- y_{0}}{x_{1} - x_{0}} $$

Finalmente obtenemos el modelo de interpolaci√≥n lineal:

$$ f(x) = y_{0} + \frac{y_{1}- y_{0}}{x_{1} - x_{0}}(x - x_{0}) $$

# 2.1.3 Polinomio de interpolaci√≥n de Newton     

Como observamos en la interpolaci√≥n lineal (ecuaci√≥n 2.2) y la interpolaci√≥n cuadr√°tica (ecuaci√≥n 2.9)   cada t√©rmino que se va aumentando es una diferencia dividida de orden mayor.  

El polinomio de interpolaci√≥n polin√≥mica para $n$ datos tiene la forma:  

$$ f(x) = a_{0} + a_{1}(x - x_{0}) + a_{2}(x-x_{0})(x-x_{1}) + \cdots + a_{n}(x - x_{0})(x - x_{1}) \cdots (x - x_{n-1}) $$

donde  

$a_{0} = y_{0}$  

$a_{1} = \frac{y_{1} - y_{0}}{x_{1} - x_{0}} = f[x_{1}, x_{0}]$

$a_{2} = \frac{\frac{y_{2} - y_{1}}{x_{2} - x_{1}} - \frac{y_{1} - y_{0}}{x_{1} - x_{0}}}{x_{2} - x_{0}} = \frac{f[x_{2}, x_{1}] - f[x_{1}, x_{0}]}{x_{2} - x_{0}} = f[x_{2}, x_{1}, x_{0}]$

$a_{n} = f[x_{n}, x_{n-1}, \ldots, x_{0}]$

La **f√≥rmula de interpolaci√≥n de Newton** en t√©rminos de **diferencias divididas** se expresa en LaTeX de la siguiente manera:



### **F√≥rmula General**:



$$ P_n(x) = f[x_0] + \sum_{k=1}^{n} \left( f[x_0, x_1, \dots, x_k] \cdot \prod_{j=0}^{k-1} (x - x_j) \right) $$




### **Explicaci√≥n de los componentes**:
1. **Diferencias divididas**:
   - **Primer orden**: 

$$ f[x_i, x_{i+1}] = \frac{f(x_{i+1}) - f(x_i)}{x_{i+1} - x_i} $$



   - **Orden superior** (recursivo): 

 
$$ f[x_i, \dots, x_{i+k}] = \frac{f[x_{i+1}, \dots, x_{i+k}] - f[x_i, \dots, x_{i+k-1}]}{x_{i+k} - x_i} $$
     



2. **T√©rminos del polinomio**:
   - El producto $\prod_{j=0}^{k-1} (x - x_j)$ representa los factores $(x - x_0)(x - x_1)\dots(x - x_{k-1})$.



### **Ejemplo completo en LaTeX**:




El polinomio de interpolaci√≥n de Newton con diferencias divididas es:

$$
P_n(x) = f[x_0] + f[x_0, x_1](x - x_0) + f[x_0, x_1, x_2](x - x_0)(x - x_1) + \dots + f[x_0, \dots, x_n]\prod_{j=0}^{n-1} (x - x_j)
$$

donde las diferencias divididas se calculan recursivamente:

$$
f[x_i, \dots, x_{i+k}] = \frac{f[x_{i+1}, \dots, x_{i+k}] - f[x_i, \dots, x_{i+k-1}]}{x_{i+k} - x_i}
$$






### **Notas**:
- **Ventaja de la forma de Newton**: Permite a√±adir nuevos puntos de interpolaci√≥n sin recalcular todo el polinomio (solo a√±adiendo un t√©rmino m√°s).
- **Uso com√∫n**: En algoritmos num√©ricos para interpolaci√≥n o construcci√≥n de tablas de diferencias divididas.

````python

# Configuraci√≥n inicial para notebook interactivo
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import ipywidgets as widgets
from scipy.interpolate import interp1d

%matplotlib inline

# Generaci√≥n de datos clim√°ticos simulados para Caucasia-Antioquia (promedios mensuales 2010-2020)
np.random.seed(42)  # Para reproducibilidad

# Meses del a√±o (1 a 12)
meses = np.arange(1, 13)

# Temperaturas promedio mensuales en ¬∞C (simuladas con patr√≥n estacional)
temp_base = 28.5  # Temperatura base anual
amplitud = 1.5    # Amplitud de variaci√≥n estacional
ruido = np.random.normal(0, 0.3, size=len(meses))  # Ruido aleatorio

# Modelo de temperatura con patr√≥n sinusoidal + ruido
temps = temp_base + amplitud * np.sin(2 * np.pi * (meses - 4) / 12) + ruido

# Ajustamos algunos valores para mayor realismo
temps[3] += 0.7  # Abril m√°s caliente
temps[7] -= 0.5  # Agosto m√°s fresco
temps = np.round(temps, 1)  # Redondeamos a 1 decimal

print("Datos clim√°ticos generados:")
print("Meses:", meses)
print("Temperaturas (¬∞C):", temps)

# Ejemplo de animaci√≥n comparativa
fig, ax = plt.subplots(figsize=(10, 6))

def update(frame):
    ax.clear()
    grado = frame + 1
    coeffs = np.polyfit(meses, temps, grado)
    modelo = np.poly1d(coeffs)
    
    ax.scatter(meses, temps, color='red', label='Datos reales')
    x_smooth = np.linspace(1, 13, 100)
    ax.plot(x_smooth, modelo(x_smooth), label=f'Regresi√≥n (grado {grado})')
    
    # Configuraci√≥n del gr√°fico
    ax.set_title('Modelado de Temperaturas en Caucasia\nEfecto del Grado en Regresi√≥n Polinomial')
    ax.set_xlabel('Mes del a√±o')
    ax.set_ylabel('Temperatura (¬∞C)')
    ax.set_xticks(meses)
    ax.set_xticklabels(['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 
                       'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend()
    
    # A√±adir informaci√≥n adicional
    r_squared = 1 - sum((temps - modelo(meses))**2)/sum((temps - np.mean(temps))**2)
    ax.text(0.02, 0.95, f'R¬≤ = {r_squared:.3f}', transform=ax.transAxes,
            bbox=dict(facecolor='white', alpha=0.8))

# Crear la animaci√≥n
ani = FuncAnimation(fig, update, frames=5, interval=1500)

# Mostrar la animaci√≥n
HTML(ani.to_jshtml())

````


## Explicaci√≥n de los datos generados:

2. **Temperaturas**: Valores simulados con:
   - Temperatura base anual de 28.5¬∞C (t√≠pica de Caucasia)
   - Variaci√≥n estacional sinusoidal (amplitud de 1.5¬∞C)
   - Ruido aleatorio normal ($\sigma = 0.3$) para simular variabilidad interanual
   - Ajustes manuales para abril (m√°s caliente) y agosto (m√°s fresco)
   - Valores redondeados a 1 decimal para mayor realismo
