## Situaci√≥n real: **Lanzamiento de un proyectil** 



Un estudiante lanza una pelota desde el suelo con una velocidad inicial $v_0$ y un √°ngulo $\theta$.
La funci√≥n de **altura** en funci√≥n del tiempo $t$ es:

$$
h(t) = v_0 \sin(\theta) \, t - \tfrac{1}{2} g t^2
$$

donde $g = 9.8 \, \text{m/s}^2$.

El **cero de esta funci√≥n** $h(t) = 0$ (despu√©s de $t=0$) corresponde al **tiempo en que la pelota vuelve a tocar el suelo**.

---



## Funci√≥n modelo

Si tomamos, por ejemplo:

* $v_0 = 20 \, \text{m/s}$
* $\theta = 45^\circ$

la funci√≥n queda:

$$
f(t) = 20 \cdot \tfrac{\sqrt{2}}{2} \, t - 4.9 t^2 = 14.14 t - 4.9 t^2
$$

---



## Interpretaci√≥n del cero

Resolver $f(t) = 0$ significa **encontrar el tiempo en que la pelota vuelve al suelo**.

Ese tiempo se obtiene anal√≠ticamente como:

$$
t = \frac{2 v_0 \sin(\theta)}{g} = \frac{2 \cdot 20 \cdot 0.707}{9.8} \approx 2.88 \, \text{segundos}
$$



Pero **si no tuvi√©ramos la f√≥rmula expl√≠cita** (por ejemplo, en un modelo m√°s complejo con resistencia del aire, terreno inclinado, etc.), ser√≠a necesario aplicar **m√©todos num√©ricos** como Newton-Raphson para aproximar el valor del cero.

---



## Ejemplo de uso de Newton-Raphson

La funci√≥n y su derivada:

$$
f(t) = 14.14 t - 4.9 t^2, \quad f'(t) = 14.14 - 9.8t
$$

Iterando con Newton a partir de una estimaci√≥n inicial $t_0 = 1$:

$$
t_{n+1} = t_n - \frac{f(t_n)}{f'(t_n)}
$$

se aproxima r√°pidamente al cero no trivial $t \approx 2.88$.

---



En resumen:

* La **funci√≥n modelo** es la altura de un proyectil.
* El **cero** representa el **tiempo en que vuelve al suelo**.
* En casos reales m√°s complejos, este problema **no tiene soluci√≥n exacta** y se necesita un **m√©todo num√©rico como Newton-Raphson**.

---

# Situaci√≥n real con resistencia del aire



Supongamos de nuevo el **lanzamiento de una pelota** con:

* Velocidad inicial $v_0$,
* √Ångulo $\theta$,
* Resistencia del aire proporcional a la velocidad (modelo lineal: fuerza de arrastre $-kv$).

---



### Ecuaciones del movimiento

En direcci√≥n **vertical**, la velocidad satisface la ecuaci√≥n diferencial:

$$
m \frac{dv}{dt} = -mg - kv
$$

donde:

* $m$ = masa del proyectil,
* $g$ = gravedad,
* $k$ = coeficiente de resistencia.



La soluci√≥n de esa ecuaci√≥n lleva a una **posici√≥n vertical** (altura):

$$
h(t) = \frac{m}{k} \left( v_0 \sin(\theta) + \frac{mg}{k} \right) \left( 1 - e^{-\tfrac{k}{m} t} \right) - \frac{mg}{k} t
$$

---



### Interpretaci√≥n

* Aqu√≠, $h(t)$ es la **altura de la pelota en el tiempo $t$**.
* El **cero de $h(t)$** (distinto de $t=0$) corresponde nuevamente al **tiempo en que la pelota toca el suelo**.



Pero ahora, **no existe una f√≥rmula cerrada** para resolver $h(t) = 0$.
El cero debe obtenerse **con un m√©todo num√©rico**, como Newton-Raphson, bisecci√≥n o secante.

---



### Ejemplo con valores

Tomemos:

* $v_0 = 20 \, \text{m/s}$,
* $\theta = 45^\circ$,
* $m = 0.15 \, \text{kg}$,
* $k = 0.1 \, \text{kg/s}$,
* $g = 9.8 \, \text{m/s}^2$.



Entonces la funci√≥n es:

$$
h(t) = \frac{0.15}{0.1} \left( 20 \cdot \tfrac{\sqrt{2}}{2} + \frac{0.15 \cdot 9.8}{0.1} \right) \left( 1 - e^{-\tfrac{0.1}{0.15} t} \right) - \frac{0.15 \cdot 9.8}{0.1} t
$$

---



### Por qu√© aqu√≠ s√≠ se justifica Newton

* En el caso **sin resistencia**, el tiempo de impacto se calcula con una f√≥rmula exacta.
* En el caso **con resistencia**, **la ecuaci√≥n involucra una exponencial** ‚Üí imposible despejar $t$ en forma cerrada.
* Necesitamos un **m√©todo num√©rico** para encontrar el cero $h(t) = 0$.

---



As√≠, la situaci√≥n se vuelve mucho m√°s **interesante y realista**, porque la resistencia del aire siempre afecta el movimiento de objetos, y el **uso de Newton-Raphson no es opcional, sino necesario**.

---



# **script en Python con Newton-Raphson** que calcule el tiempo en que la pelota toca el suelo considerando resistencia del aire, y adem√°s un gr√°fico de la trayectoria



Aqu√≠ tienes un **script en Python** que define y grafica la funci√≥n de altura $h(t)$ con resistencia del aire usando **matplotlib**.



He usado los par√°metros de ejemplo que te propuse:

* $v_0 = 20 \, \text{m/s}$
* $\theta = 45^\circ$
* $m = 0.15 \, \text{kg}$
* $k = 0.1 \, \text{kg/s}$
* $g = 9.8 \, \text{m/s}^2$


````python
import numpy as np
import matplotlib.pyplot as plt
````

````python

# Par√°metros f√≠sicos
v0 = 20.0        # velocidad inicial (m/s)
theta = np.pi/4  # √°ngulo de lanzamiento (rad)
m = 0.15         # masa (kg)
k = 0.1          # coeficiente de resistencia (kg/s)
g = 10           # gravedad (m/s^2)
````

````python

# Funci√≥n altura con resistencia del aire
def h(t):
    term1 = (m/k) * (v0*np.sin(theta) + (m*g)/k) * (1 - np.exp(-k*t/m))
    term2 = (m*g/k) * t
    return term1 - term2
````

# implementaci√≥n del **m√©todo de Newton-Raphson** para encontrar un cero de la funci√≥n $h(t)$, que representa la **altura con resistencia del aire**.   

Te preparo un script en **Python** que:

1. Define la funci√≥n $h(t)$.
2. Calcula su derivada de forma num√©rica.
3. Implementa el m√©todo de Newton.
4. Devuelve las iteraciones en un **DataFrame** con columnas: `iteraci√≥n`, `t_aprox`, `h(t)`, `error`.



Aqu√≠ va el c√≥digo:



````python

import numpy as np
import pandas as pd
````

````python

# Par√°metros f√≠sicos
m = 0.15   # masa (kg) de una pelota de b√©isbol aprox
v0 = 20.0   # velocidad inicial (m/s)
theta = np.pi/4  # √°ngulo de lanzamiento en radianes
g = 9.8    # gravedad (m/s^2)
k = 0.1     # coeficiente de resistencia del aire

# Definici√≥n de la funci√≥n h(t)
def h(t):
    term1 = (m/k) * (v0*np.sin(theta) + (m*g)/k) * (1 - np.exp(-k*t/m))
    term2 = (m*g/k) * t
    return term1 - term2

# Derivada num√©rica de h(t)
def h_prime(t, h=h, h_step=1e-5):
    return (h(t + h_step) - h(t - h_step)) / (2*h_step)

# M√©todo de Newton-Raphson
def newton_method(f, f_prime, x0, tol=1e-6, max_iter=20):
    resultados = []
    x = x0
    for i in range(max_iter):
        fx = f(x)
        fpx = f_prime(x)
        if fpx == 0:
            print("Derivada nula. M√©todo detenido.")
            break
        x_new = x - fx/fpx # implementaci√≥n de la f√≥rmula de Newton
        error = abs(x_new - x)
        resultados.append([i+1, x_new, f(x_new), error])
        if error < tol:
            break
        x = x_new
    df = pd.DataFrame(resultados, columns=["Iteraci√≥n", "t_aprox", "h(t)", "Error"])
    return df
````

````python

# Ejemplo de uso
t0 = 4  # valor inicial
df_resultados = newton_method(h, h_prime, t0)

print("Resultados m√©todo de Newton\n", df_resultados)

````

````python
def newton_raphson(f, df, x0, epsilon=0.000000001, max_iter=100):
    """
    Encuentra la ra√≠z de f(x) = 0 con precisi√≥n epsilon.
    
    Par√°metros:
        f: Funci√≥n a resolver.
        df: Derivada de f.
        x0: Punto inicial.
        epsilon: Precisi√≥n deseada.
        max_iter: M√°ximo de iteraciones.
    
    Retorna:
        (ra√≠z, iteraciones, historia)
    """
    historia = []
    for n in range(max_iter):
        x1 = x0 - f(x0)/df(x0)
        historia.append(x1)
        if abs(x1 - x0) < epsilon:
            return historia
        x0 = x1
    raise ValueError("No converge en {} iteraciones".format(max_iter))
````

# Error Relativo en el M√©todo de Newton

El error relativo es una medida importante para evaluar la convergencia en el M√©todo de Newton (tambi√©n conocido como Newton-Raphson). 

Este indicador nos permite determinar cu√°ndo detener las iteraciones al alcanzar una precisi√≥n satisfactoria.



## Definici√≥n del Error Relativo

El error relativo (Œµ) en la iteraci√≥n k+1 se define como:


$$ \epsilon = |x_{k+1} - x_k| / |x_{k+1}| $$


Donde:
- `x_k` es la aproximaci√≥n actual
- `x_{k+1}` es la siguiente aproximaci√≥n



## Implementaci√≥n en Python


````python


def newton_raphson(f, df, x0, tol=1e-6, max_iter=100):
    """
    Implementaci√≥n del M√©todo de Newton-Raphson con criterio de parada por error relativo
    
    Par√°metros:
    f - funci√≥n a encontrar la ra√≠z
    df - derivada de la funci√≥n f
    x0 - aproximaci√≥n inicial
    tol - tolerancia (error relativo aceptable)
    max_iter - n√∫mero m√°ximo de iteraciones
    
    Retorna:
    Una tupla con (ra√≠z aproximada, n√∫mero de iteraciones, historial de errores)
    """
    errors = []
    x_prev = x0
    
    for i in range(max_iter):
        x_next = x_prev - f(x_prev)/df(x_prev)
        
        # Calcular error relativo
        if x_next != 0:
            rel_error = abs(x_next - x_prev)/abs(x_next)
        else:
            rel_error = float('inf')
            
        errors.append(rel_error)
        
        # Verificar criterio de convergencia
        if rel_error < tol:
            break
            
        x_prev = x_next
    
    return x_next, i+1, errors

````


## Ejemplo de Uso


````python


import math

# Definir funci√≥n y su derivada
def f(x):
    return x**3 - 2*x - 5

def df(x):
    return 3*x**2 - 2

# Aplicar m√©todo de Newton
raiz, iteraciones, errores = newton_raphson(f, df, x0=3.0, tol=1e-8)

print(f"Ra√≠z encontrada: {raiz:.8f}")
print(f"Iteraciones realizadas: {iteraciones}")
print(f"Error relativo final: {errores[-1]:.2e}")

# Graficar convergencia del error
import matplotlib.pyplot as plt

plt.semilogy(range(iteraciones), errores, 'o-')
plt.title('Convergencia del Error Relativo en M√©todo de Newton')
plt.xlabel('Iteraci√≥n')
plt.ylabel('Error Relativo')
plt.grid(True)
plt.show()

````


## Interpretaci√≥n del Error Relativo

1. **Convergencia cuadr√°tica**: El m√©todo de Newton t√≠picamente muestra convergencia cuadr√°tica cerca de la ra√≠z, lo que significa que el error relativo disminuye r√°pidamente.

2. **Criterio de parada**: El algoritmo se detiene cuando el error relativo es menor que la tolerancia especificada (`tol`).

3. **Estabilidad num√©rica**: El error relativo es m√°s robusto que el error absoluto para valores cercanos a cero.

## Consideraciones Adicionales

- Para evitar divisi√≥n por cero, se debe manejar el caso cuando `x_next = 0`
- En algunas implementaciones se usa un error relativo modificado: `Œµ = |x_{k+1} - x_k| / (|x_{k+1}| + Œ¥)` donde Œ¥ es una peque√±a constante para mejorar estabilidad
- El error relativo puede no ser adecuado cuando la soluci√≥n es exactamente cero

El error relativo proporciona una medida escalada de la convergencia, haciendo que el criterio de parada sea m√°s consistente para problemas con magnitudes variables.

````python
import matplotlib.pyplot as plt
import numpy as np

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import matplotlib

# Configuraci√≥n espec√≠fica para VS Code
matplotlib.rcParams['animation.html'] = 'jshtml'
%matplotlib inline

# Definimos la funci√≥n y su derivada
f = lambda x: np.exp(x) + x  # Funci√≥n ejemplo: e^x + x
df = lambda x: np.exp(x) + 1  # Su derivada

def animate_newton(f, df, x0, epsilon, max_iter=10):
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Rango de visualizaci√≥n adaptativo
    x_min, x_max = x0-0.5, x0+0.5
    x_vals = np.linspace(x_min, x_max, 400)
    
    # Graficar la funci√≥n y el eje x
    ax.plot(x_vals, f(x_vals), label=f'f(x) = $e^x + x$', color='blue')
    ax.axhline(0, color='black', linestyle='--', linewidth=0.8)
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.set_title('M√©todo de Newton-Raphson')
    ax.grid(True, alpha=0.3)
    ax.legend()
    
    # Inicializar variables para la animaci√≥n
    points, = ax.plot([], [], 'ro', label='Aproximaciones')
    lines, = ax.plot([], [], 'r--', linewidth=1.5, alpha=0.7)
    
    # Lista para almacenar el historial de iteraciones
    history = []
    
    def init():
        points.set_data([], [])
        lines.set_data([], [])
        return points, lines
    
    def update(frame):
        nonlocal x0  # Permite modificar x0 en el scope exterior
        
        # Calcular nueva aproximaci√≥n
        x1 = x0 - f(x0)/df(x0)
        
        # Almacenar el paso actual
        history.append((x0, x1))
        
        # Actualizar gr√°ficos
        x_points = [x for x, _ in history]
        y_points = [f(x) for x, _ in history]
        points.set_data(x_points, y_points)
        
        # Dibujar l√≠neas tangentes
        if len(history) > 0:
            tangent_x = np.linspace(x0-0.5, x0+0.5, 2)
            tangent_y = f(x0) + df(x0)*(tangent_x - x0)
            lines.set_data(tangent_x, tangent_y)
        
        # Actualizar x0 para la siguiente iteraci√≥n
        x0 = x1
        
        # Ajustar l√≠mites del gr√°fico si es necesario
        ax.set_xlim(min(x_vals.min(), x1-1), max(x_vals.max(), x1+1))
        
        return points, lines
    
    # Crear la animaci√≥n
    anim = FuncAnimation(fig, update, frames=range(max_iter),
                        init_func=init, blit=True, interval=800,
                        repeat=False)
    
    # Cerrar la figura para evitar duplicados en VS Code
    plt.close()
    
    return anim

# Ejemplo de uso
animation = animate_newton(f, df, x0=-1.0, epsilon=1e-6, max_iter=5)
HTML(animation.to_jshtml())
````

**Software:** Python (SymPy para derivadas).  



````python
from math import gcd

def posibles_ceros_racionales(a, b, c, d):
    """
    Para el polinomio f(x) = a*x^3 + b*x^2 + c*x + d,
    retorna la lista de posibles ceros racionales seg√∫n el teorema de los ceros racionales.
    """
    def divisores(n):
        n = abs(n)
        divs = set()
        for i in range(1, n+1):
            if n % i == 0:
                divs.add(i)
                divs.add(-i)
        return divs

    p_divs = divisores(d)  # divisores del t√©rmino independiente
    q_divs = divisores(a)  # divisores del coeficiente l√≠der

    posibles = set()
    for p in p_divs:
        for q in q_divs:
            if q == 0:
                continue
            g = gcd(p, q)
            num = p // g
            den = q // g
            posibles.add(num / den)
    return sorted(posibles)

# Ejemplo de uso para f(x) = x^3 - 2x^2 - 5x + 6
ceros = posibles_ceros_racionales(1, +2, -5, 6)
print(ceros)
````

# An√°lisis de Convergencia del M√©todo de Newton para Ra√≠ces M√∫ltiples



El m√©todo de Newton tiene convergencia cuadr√°tica para ra√≠ces simples, pero su comportamiento cambia para ra√≠ces m√∫ltiples (donde la funci√≥n toca pero no cruza el eje x). 

En este caso, la convergencia se vuelve lineal y la derivada tiende a cero cerca de la ra√≠z.



## Dise√±o del An√°lisis

1. **Definir una funci√≥n con ra√≠z m√∫ltiple**: f(x) = (x - Œ±)^m * g(x), donde m es la multiplicidad
2. **Implementar el m√©todo de Newton**
3. **Analizar la convergencia para diferentes multiplicidades**
4. **Visualizar la funci√≥n y el proceso iterativo**
5. **Comparar con el m√©todo de Newton modificado para ra√≠ces m√∫ltiples**


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton
````

````python

def funcion_con_raiz_multiple(x, multiplicidad=2, alpha=1.0):
    """Funci√≥n con ra√≠z m√∫ltiple en x=alpha"""
    return (x - alpha)**multiplicidad * np.cos(x - alpha)
````

````python
# grafique la funci√≥n cooon raiz multiple 
x = np.linspace(-2, 4, 400)
y = funcion_con_raiz_multiple(x, multiplicidad=2, alpha=1.0)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = (x - 1)^2 * cos(x - 1)', color='blue')
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.axvline(1, color='red', linestyle='--', label='Ra√≠z m√∫ltiple en x=1')
plt.title('Funci√≥n con Ra√≠z M√∫ltiple')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
save_fig('funcion_con_raiz_multiple')
plt.show()
````

````python

def derivada_con_raiz_multiple(x, multiplicidad=2, alpha=1.0):
    """Derivada de la funci√≥n con ra√≠z m√∫ltiple"""
    return multiplicidad*(x-alpha)**(multiplicidad-1)*np.cos(x-alpha) - (x-alpha)**multiplicidad*np.sin(x-alpha)
````

````python

def metodo_newton(f, df, x0, tol=1e-10, max_iter=100):
    """Implementaci√≥n b√°sica del m√©todo de Newton"""
    iteraciones = [x0]
    for _ in range(max_iter):
        x_next = x0 - f(x0)/df(x0)
        iteraciones.append(x_next)
        if abs(x_next - x0) < tol:
            break
        x0 = x_next
    return np.array(iteraciones)
````

````python

def newton_modificado(f, df, multiplicidad, x0, tol=1e-10, max_iter=100):
    """M√©todo de Newton modificado para ra√≠ces m√∫ltiples"""
    iteraciones = [x0]
    for _ in range(max_iter):
        x_next = x0 - multiplicidad * f(x0)/df(x0)
        iteraciones.append(x_next)
        if abs(x_next - x0) < tol:
            break
        x0 = x_next
    return np.array(iteraciones)
````

````python

def analizar_convergencia(multiplicidad=2):
    """An√°lisis completo para una multiplicidad dada"""
    alpha = 2.0
    f = lambda x: funcion_con_raiz_multiple(x, multiplicidad, alpha)
    df = lambda x: derivada_con_raiz_multiple(x, multiplicidad, alpha)
    
    # Punto inicial cerca de la ra√≠z
    x0 = alpha + 0.5
    
    # Aplicar m√©todo de Newton est√°ndar
    iter_newton = metodo_newton(f, df, x0)
    
    # Aplicar m√©todo de Newton modificado
    iter_modificado = newton_modificado(f, df, multiplicidad, x0)
    
    # Calcular errores
    errores_newton = np.abs(iter_newton - alpha)
    errores_modificado = np.abs(iter_modificado - alpha)
    
    # Graficar funci√≥n y derivada
    plt.figure(figsize=(12, 10))
    
    # Gr√°fico de la funci√≥n
    plt.subplot(2, 2, 1)
    x_vals = np.linspace(alpha-2, alpha+2, 400)
    plt.plot(x_vals, f(x_vals), label=f'f(x) = (x-{alpha})^{multiplicidad}*cos(x-{alpha})')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(alpha, color='red', linestyle='--', label=f'Ra√≠z en x={alpha}')
    plt.title(f'Funci√≥n con ra√≠z de multiplicidad {multiplicidad}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    
    # Gr√°fico de la derivada
    plt.subplot(2, 2, 2)
    plt.plot(x_vals, df(x_vals), label=f"f'(x)")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(alpha, color='red', linestyle='--')
    plt.title('Derivada de la funci√≥n')
    plt.xlabel('x')
    plt.ylabel("f'(x)")
    plt.legend()
    plt.grid(True)
    
    # Gr√°fico de convergencia
    plt.subplot(2, 2, 3)
    plt.semilogy(errores_newton, 'o-', label='Newton est√°ndar')
    plt.semilogy(errores_modificado, 's-', label=f'Newton modificado (m={multiplicidad})')
    plt.title('Convergencia de los m√©todos')
    plt.xlabel('Iteraci√≥n')
    plt.ylabel('Error absoluto (escala log)')
    plt.legend()
    plt.grid(True)
    
    # Gr√°fico de relaci√≥n de convergencia
    plt.subplot(2, 2, 4)
    if len(errores_newton) > 1:
        ratios_newton = errores_newton[1:]/errores_newton[:-1]
        plt.plot(ratios_newton, 'o-', label='Ratio Newton')
    if len(errores_modificado) > 1:
        ratios_modificado = errores_modificado[1:]/errores_modificado[:-1]**2
        plt.plot(ratios_modificado, 's-', label='Ratio Newton modificado')
    plt.title('Ratio de convergencia')
    plt.xlabel('Iteraci√≥n')
    plt.ylabel('Ratio de error')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    # Mostrar resultados num√©ricos
    print(f"\nAn√°lisis para multiplicidad = {multiplicidad}")
    print("Newton est√°ndar:")
    print(f"Iteraciones: {len(iter_newton)}, Error final: {errores_newton[-1]:.2e}")
    print("\nNewton modificado:")
    print(f"Iteraciones: {len(iter_modificado)}, Error final: {errores_modificado[-1]:.2e}")
````

# Explicaci√≥n del Ratio de Convergencia en el An√°lisis del M√©todo de Newton



El ratio de convergencia es una medida fundamental para entender c√≥mo de r√°pido un m√©todo iterativo se aproxima a la soluci√≥n exacta. En el contexto del m√©todo de Newton para ra√≠ces m√∫ltiples, este concepto es especialmente importante para caracterizar el comportamiento del algoritmo.



## Concepto Matem√°tico del Ratio de Convergencia



Dada una sucesi√≥n $\{x_{n}\}$ que converge a Œ±, definimos el error en la iteraci√≥n n como:
e‚Çô = |x‚Çô - Œ±|

El **ratio de convergencia** en cada paso se calcula como:
R‚Çô = e‚Çô‚Çä‚ÇÅ / e‚Çô^p

Donde p es el **orden de convergencia** que queremos analizar:
- p=1 para convergencia lineal
- p=2 para convergencia cuadr√°tica



## Comportamiento en el C√≥digo Anterior

En el c√≥digo proporcionado, calculamos dos tipos de ratios:

1. **Para el m√©todo de Newton est√°ndar**:
   ```python
   ratios_newton = errores_newton[1:]/errores_newton[:-1]  # p=1
   ```
   Esto calcula R‚Çô = e‚Çô‚Çä‚ÇÅ/e‚Çô (convergencia lineal)

2. **Para el m√©todo de Newton modificado**:
   ```python
   ratios_modificado = errores_modificado[1:]/errores_modificado[:-1]**2  # p=2
   ```
   Esto calcula R‚Çô = e‚Çô‚Çä‚ÇÅ/e‚Çô¬≤ (convergencia cuadr√°tica)

## Interpretaci√≥n para Ra√≠ces M√∫ltiples

Para una ra√≠z de multiplicidad m:

1. **M√©todo de Newton est√°ndar**:
   - Orden de convergencia: 1 (lineal)
   - Ratio te√≥rico: R ‚Üí (m-1)/m cuando n ‚Üí ‚àû
   - Ejemplo para m=2: R ‚Üí 0.5

2. **M√©todo de Newton modificado**:
   - Orden de convergencia: 2 (cuadr√°tico)
   - Ratio te√≥rico: R ‚Üí constante (convergencia m√°s r√°pida)
