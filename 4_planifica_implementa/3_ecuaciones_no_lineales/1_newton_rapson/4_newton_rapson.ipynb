## Situación real: **Lanzamiento de un proyectil** 



Un estudiante lanza una pelota desde el suelo con una velocidad inicial $v_0$ y un ángulo $\theta$.
La función de **altura** en función del tiempo $t$ es:

$$
h(t) = v_0 \sin(\theta) \, t - \tfrac{1}{2} g t^2
$$

donde $g = 9.8 \, \text{m/s}^2$.

El **cero de esta función** $h(t) = 0$ (después de $t=0$) corresponde al **tiempo en que la pelota vuelve a tocar el suelo**.

---



## Función modelo

Si tomamos, por ejemplo:

* $v_0 = 20 \, \text{m/s}$
* $\theta = 45^\circ$

la función queda:

$$
f(t) = 20 \cdot \tfrac{\sqrt{2}}{2} \, t - 4.9 t^2 = 14.14 t - 4.9 t^2
$$

---



## Interpretación del cero

Resolver $f(t) = 0$ significa **encontrar el tiempo en que la pelota vuelve al suelo**.

Ese tiempo se obtiene analíticamente como:

$$
t = \frac{2 v_0 \sin(\theta)}{g} = \frac{2 \cdot 20 \cdot 0.707}{9.8} \approx 2.88 \, \text{segundos}
$$



Pero **si no tuviéramos la fórmula explícita** (por ejemplo, en un modelo más complejo con resistencia del aire, terreno inclinado, etc.), sería necesario aplicar **métodos numéricos** como Newton-Raphson para aproximar el valor del cero.

---



## Ejemplo de uso de Newton-Raphson

La función y su derivada:

$$
f(t) = 14.14 t - 4.9 t^2, \quad f'(t) = 14.14 - 9.8t
$$

Iterando con Newton a partir de una estimación inicial $t_0 = 1$:

$$
t_{n+1} = t_n - \frac{f(t_n)}{f'(t_n)}
$$

se aproxima rápidamente al cero no trivial $t \approx 2.88$.

---



En resumen:

* La **función modelo** es la altura de un proyectil.
* El **cero** representa el **tiempo en que vuelve al suelo**.
* En casos reales más complejos, este problema **no tiene solución exacta** y se necesita un **método numérico como Newton-Raphson**.

---

# Situación real con resistencia del aire



Supongamos de nuevo el **lanzamiento de una pelota** con:

* Velocidad inicial $v_0$,
* Ángulo $\theta$,
* Resistencia del aire proporcional a la velocidad (modelo lineal: fuerza de arrastre $-kv$).

---



### Ecuaciones del movimiento

En dirección **vertical**, la velocidad satisface la ecuación diferencial:

$$
m \frac{dv}{dt} = -mg - kv
$$

donde:

* $m$ = masa del proyectil,
* $g$ = gravedad,
* $k$ = coeficiente de resistencia.



La solución de esa ecuación lleva a una **posición vertical** (altura):

$$
h(t) = \frac{m}{k} \left( v_0 \sin(\theta) + \frac{mg}{k} \right) \left( 1 - e^{-\tfrac{k}{m} t} \right) - \frac{mg}{k} t
$$

---



### Interpretación

* Aquí, $h(t)$ es la **altura de la pelota en el tiempo $t$**.
* El **cero de $h(t)$** (distinto de $t=0$) corresponde nuevamente al **tiempo en que la pelota toca el suelo**.



Pero ahora, **no existe una fórmula cerrada** para resolver $h(t) = 0$.
El cero debe obtenerse **con un método numérico**, como Newton-Raphson, bisección o secante.

---



### Ejemplo con valores

Tomemos:

* $v_0 = 20 \, \text{m/s}$,
* $\theta = 45^\circ$,
* $m = 0.15 \, \text{kg}$,
* $k = 0.1 \, \text{kg/s}$,
* $g = 9.8 \, \text{m/s}^2$.



Entonces la función es:

$$
h(t) = \frac{0.15}{0.1} \left( 20 \cdot \tfrac{\sqrt{2}}{2} + \frac{0.15 \cdot 9.8}{0.1} \right) \left( 1 - e^{-\tfrac{0.1}{0.15} t} \right) - \frac{0.15 \cdot 9.8}{0.1} t
$$

---



### Por qué aquí sí se justifica Newton

* En el caso **sin resistencia**, el tiempo de impacto se calcula con una fórmula exacta.
* En el caso **con resistencia**, **la ecuación involucra una exponencial** → imposible despejar $t$ en forma cerrada.
* Necesitamos un **método numérico** para encontrar el cero $h(t) = 0$.

---



Así, la situación se vuelve mucho más **interesante y realista**, porque la resistencia del aire siempre afecta el movimiento de objetos, y el **uso de Newton-Raphson no es opcional, sino necesario**.

---

````python

# Parámetros físicos
v0 = 20.0        # velocidad inicial (m/s)
theta = np.pi/4  # ángulo de lanzamiento (rad)
m = 0.15         # masa (kg)
k = 0.1          # coeficiente de resistencia (kg/s)
g = 10           # gravedad (m/s^2)
````

````python

# Función altura con resistencia del aire
def h(t):
    term1 = (m/k) * (v0*np.sin(theta) + (m*g)/k) * (1 - np.exp(-k*t/m))
    term2 = (m*g/k) * t
    return term1 - term2
````

# implementación del **método de Newton-Raphson** para encontrar un cero de la función $h(t)$, que representa la **altura con resistencia del aire**.   

Te preparo un script en **Python** que:

1. Define la función $h(t)$.
2. Calcula su derivada de forma numérica.
3. Implementa el método de Newton.
4. Devuelve las iteraciones en un **DataFrame** con columnas: `iteración`, `t_aprox`, `h(t)`, `error`.



Aquí va el código:



````python

import numpy as np
import pandas as pd
````

````python

# Parámetros físicos
m = 0.15   # masa (kg) de una pelota de béisbol aprox
v0 = 20.0   # velocidad inicial (m/s)
theta = np.pi/4  # ángulo de lanzamiento en radianes
g = 9.8    # gravedad (m/s^2)
k = 0.1     # coeficiente de resistencia del aire

# Definición de la función h(t)
def h(t):
    term1 = (m/k) * (v0*np.sin(theta) + (m*g)/k) * (1 - np.exp(-k*t/m))
    term2 = (m*g/k) * t
    return term1 - term2

# Derivada numérica de h(t)
def h_prime(t, h=h, h_step=1e-5):
    return (h(t + h_step) - h(t - h_step)) / (2*h_step)

# Método de Newton-Raphson
def newton_method(f, f_prime, x0, tol=1e-6, max_iter=20):
    resultados = []
    x = x0
    for i in range(max_iter):
        fx = f(x)
        fpx = f_prime(x)
        if fpx == 0:
            print("Derivada nula. Método detenido.")
            break
        x_new = x - fx/fpx # implementación de la fórmula de Newton
        error = abs(x_new - x)
        resultados.append([i+1, x_new, f(x_new), error])
        if error < tol:
            break
        x = x_new
    df = pd.DataFrame(resultados, columns=["Iteración", "t_aprox", "h(t)", "Error"])
    return df
````

````python

# Ejemplo de uso
t0 = 4  # valor inicial
df_resultados = newton_method(h, h_prime, t0)

print("Resultados método de Newton\n", df_resultados)

````

# Contemos de donde puede venir este problema  



### **Deducción del Modelo Matemático para Optimización de Riego en Cultivos de Arroz**  


**Contexto real en Antioquia, Colombia**:  
En la región de **El Bajo Cauca antioqueño**, los ingenieros agropecuarios enfrentan el desafío de determinar el **volumen óptimo de agua** (en m³/ha) para cultivos de arroz, donde el exceso o defecto de riego afecta la productividad.  





#### **1. Situación Real**  
Supongamos que:  
- **La tasa de crecimiento del arroz** (en kg/ha) sigue una relación cúbica con el volumen de agua aplicada ($x$).  
- **Datos empíricos** muestran que:  
  - Con $x = 1$ m³/ha → Crecimiento = $-6$ kg/ha (déficit hídrico).  
  - Con $x = 2$ m³/ha → Crecimiento = \(-1\) kg/ha (aún insuficiente).  
  - Con $x = 3$ m³/ha → Crecimiento = \(16\) kg/ha (óptimo).  



**Problema**:  
Encontrar el volumen exacto $x$ donde el crecimiento es **cero** (punto de transición entre déficit y crecimiento positivo), para ajustar el riego sin desperdiciar agua.  





#### **2. Construcción del Modelo Matemático**  
Se propone una función cúbica que modele el crecimiento:  
$$
f(x) = x^3 - 2x - 5
$$  



**¿Por qué esta función?**  
- **Condiciones dadas**:  
  1. $f(1) = 1^3 - 2(1) - 5 = -6$ ✔️  
  2. $f(2) = 8 - 4 - 5 = -1$ ✔️  
  3. $f(3) = 27 - 6 - 5 = 16$ ✔️  



- **Interpretación**:  
  - La raíz de $f(x) = 0$ ($x \approx 2.0946$) indica el **volumen crítico** donde el crecimiento pasa de negativo a positivo.  





#### **3. Deducción Gráfica**  


````python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 3, 100)
f = lambda x: x**3 - 2*x - 5

plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label='$f(x) = x^3 - 2x - 5$', color='blue')
plt.axhline(0, color='black', linestyle='--')
plt.scatter([2.0946], [0], color='red', label='Raíz ($x \\approx 2.0946$)')
plt.xlabel('Volumen de agua (m³/ha)')
plt.ylabel('Crecimiento del arroz (kg/ha)')
plt.title('Modelo de Crecimiento vs. Riego en Arroz')
plt.legend()
plt.grid(True)
plt.show()
````

 
**Salida**:  
- La gráfica muestra que la raíz está cerca de \(x = 2.0946\), el punto exacto donde el crecimiento es cero.  





#### **4. Relación con el Método Newton-Raphson**  
**Aplicación**:  
Para encontrar numéricamente la raíz:  
1. **Derivada**: $f'(x) = 3x^2 - 2$.  
2. **Iteración**:  
   $$
   x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
   $$  
   - Con $x_0 = 2$:  
     - $x_1 = 2 - \frac{-1}{10} = 2.1$  
     - $x_2 = 2.1 - \frac{0.061}{11.23} \approx 2.0946$  



**Resultado**:

````python
def newton_raphson(f, df, x0, epsilon=0.000000001, max_iter=100):
    """
    Encuentra la raíz de f(x) = 0 con precisión epsilon.
    
    Parámetros:
        f: Función a resolver.
        df: Derivada de f.
        x0: Punto inicial.
        epsilon: Precisión deseada.
        max_iter: Máximo de iteraciones.
    
    Retorna:
        (raíz, iteraciones, historia)
    """
    historia = []
    for n in range(max_iter):
        x1 = x0 - f(x0)/df(x0)
        historia.append(x1)
        if abs(x1 - x0) < epsilon:
            return historia
        x0 = x1
    raise ValueError("No converge en {} iteraciones".format(max_iter))
````

# Error Relativo en el Método de Newton

El error relativo es una medida importante para evaluar la convergencia en el Método de Newton (también conocido como Newton-Raphson). 

Este indicador nos permite determinar cuándo detener las iteraciones al alcanzar una precisión satisfactoria.



## Definición del Error Relativo

El error relativo (ε) en la iteración k+1 se define como:


$$ \epsilon = |x_{k+1} - x_k| / |x_{k+1}| $$


Donde:
- `x_k` es la aproximación actual
- `x_{k+1}` es la siguiente aproximación



## Implementación en Python


````python


def newton_raphson(f, df, x0, tol=1e-6, max_iter=100):
    """
    Implementación del Método de Newton-Raphson con criterio de parada por error relativo
    
    Parámetros:
    f - función a encontrar la raíz
    df - derivada de la función f
    x0 - aproximación inicial
    tol - tolerancia (error relativo aceptable)
    max_iter - número máximo de iteraciones
    
    Retorna:
    Una tupla con (raíz aproximada, número de iteraciones, historial de errores)
    """
    errors = []
    x_prev = x0
    
    for i in range(max_iter):
        x_next = x_prev - f(x_prev)/df(x_prev)
        
        # Calcular error relativo
        if x_next != 0:
            rel_error = abs(x_next - x_prev)/abs(x_next)
        else:
            rel_error = float('inf')
            
        errors.append(rel_error)
        
        # Verificar criterio de convergencia
        if rel_error < tol:
            break
            
        x_prev = x_next
    
    return x_next, i+1, errors

````


## Ejemplo de Uso


````python


import math

# Definir función y su derivada
def f(x):
    return x**3 - 2*x - 5

def df(x):
    return 3*x**2 - 2

# Aplicar método de Newton
raiz, iteraciones, errores = newton_raphson(f, df, x0=3.0, tol=1e-8)

print(f"Raíz encontrada: {raiz:.8f}")
print(f"Iteraciones realizadas: {iteraciones}")
print(f"Error relativo final: {errores[-1]:.2e}")

# Graficar convergencia del error
import matplotlib.pyplot as plt

plt.semilogy(range(iteraciones), errores, 'o-')
plt.title('Convergencia del Error Relativo en Método de Newton')
plt.xlabel('Iteración')
plt.ylabel('Error Relativo')
plt.grid(True)
plt.show()

````


## Interpretación del Error Relativo

1. **Convergencia cuadrática**: El método de Newton típicamente muestra convergencia cuadrática cerca de la raíz, lo que significa que el error relativo disminuye rápidamente.

2. **Criterio de parada**: El algoritmo se detiene cuando el error relativo es menor que la tolerancia especificada (`tol`).

3. **Estabilidad numérica**: El error relativo es más robusto que el error absoluto para valores cercanos a cero.

## Consideraciones Adicionales

- Para evitar división por cero, se debe manejar el caso cuando `x_next = 0`
- En algunas implementaciones se usa un error relativo modificado: `ε = |x_{k+1} - x_k| / (|x_{k+1}| + δ)` donde δ es una pequeña constante para mejorar estabilidad
- El error relativo puede no ser adecuado cuando la solución es exactamente cero

El error relativo proporciona una medida escalada de la convergencia, haciendo que el criterio de parada sea más consistente para problemas con magnitudes variables.

````python
import matplotlib.pyplot as plt
import numpy as np

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import matplotlib

# Configuración específica para VS Code
matplotlib.rcParams['animation.html'] = 'jshtml'
%matplotlib inline

# Definimos la función y su derivada
f = lambda x: np.exp(x) + x  # Función ejemplo: e^x + x
df = lambda x: np.exp(x) + 1  # Su derivada

def animate_newton(f, df, x0, epsilon, max_iter=10):
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Rango de visualización adaptativo
    x_min, x_max = x0-0.5, x0+0.5
    x_vals = np.linspace(x_min, x_max, 400)
    
    # Graficar la función y el eje x
    ax.plot(x_vals, f(x_vals), label=f'f(x) = $e^x + x$', color='blue')
    ax.axhline(0, color='black', linestyle='--', linewidth=0.8)
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.set_title('Método de Newton-Raphson')
    ax.grid(True, alpha=0.3)
    ax.legend()
    
    # Inicializar variables para la animación
    points, = ax.plot([], [], 'ro', label='Aproximaciones')
    lines, = ax.plot([], [], 'r--', linewidth=1.5, alpha=0.7)
    
    # Lista para almacenar el historial de iteraciones
    history = []
    
    def init():
        points.set_data([], [])
        lines.set_data([], [])
        return points, lines
    
    def update(frame):
        nonlocal x0  # Permite modificar x0 en el scope exterior
        
        # Calcular nueva aproximación
        x1 = x0 - f(x0)/df(x0)
        
        # Almacenar el paso actual
        history.append((x0, x1))
        
        # Actualizar gráficos
        x_points = [x for x, _ in history]
        y_points = [f(x) for x, _ in history]
        points.set_data(x_points, y_points)
        
        # Dibujar líneas tangentes
        if len(history) > 0:
            tangent_x = np.linspace(x0-0.5, x0+0.5, 2)
            tangent_y = f(x0) + df(x0)*(tangent_x - x0)
            lines.set_data(tangent_x, tangent_y)
        
        # Actualizar x0 para la siguiente iteración
        x0 = x1
        
        # Ajustar límites del gráfico si es necesario
        ax.set_xlim(min(x_vals.min(), x1-1), max(x_vals.max(), x1+1))
        
        return points, lines
    
    # Crear la animación
    anim = FuncAnimation(fig, update, frames=range(max_iter),
                        init_func=init, blit=True, interval=800,
                        repeat=False)
    
    # Cerrar la figura para evitar duplicados en VS Code
    plt.close()
    
    return anim

# Ejemplo de uso
animation = animate_newton(f, df, x0=-1.0, epsilon=1e-6, max_iter=5)
HTML(animation.to_jshtml())
````

3. Comparar con la solución analítica, es decir, comparar con la solución obtenida utilizando el teorema de los posibles ceros racionales, el teorema del residuo y el teorema del factor.  
**Software:** Python (SymPy para derivadas).  



````python
from math import gcd

def posibles_ceros_racionales(a, b, c, d):
    """
    Para el polinomio f(x) = a*x^3 + b*x^2 + c*x + d,
    retorna la lista de posibles ceros racionales según el teorema de los ceros racionales.
    """
    def divisores(n):
        n = abs(n)
        divs = set()
        for i in range(1, n+1):
            if n % i == 0:
                divs.add(i)
                divs.add(-i)
        return divs

    p_divs = divisores(d)  # divisores del término independiente
    q_divs = divisores(a)  # divisores del coeficiente líder

    posibles = set()
    for p in p_divs:
        for q in q_divs:
            if q == 0:
                continue
            g = gcd(p, q)
            num = p // g
            den = q // g
            posibles.add(num / den)
    return sorted(posibles)

# Ejemplo de uso para f(x) = x^3 - 2x^2 - 5x + 6
ceros = posibles_ceros_racionales(1, +2, -5, 6)
print(ceros)
````

````python
def ceros_racionales_verdaderos(f, posibles_ceros, tol=1e-8):
    ceros_verdaderos = []
    for x in posibles_ceros:
        try:
            if abs(f(x)) < tol:
                ceros_verdaderos.append(x)
        except Exception:
            continue
    return ceros_verdaderos
````

# Análisis de Convergencia del Método de Newton para Raíces Múltiples



El método de Newton tiene convergencia cuadrática para raíces simples, pero su comportamiento cambia para raíces múltiples (donde la función toca pero no cruza el eje x). 

En este caso, la convergencia se vuelve lineal y la derivada tiende a cero cerca de la raíz.



## Diseño del Análisis

1. **Definir una función con raíz múltiple**: f(x) = (x - α)^m * g(x), donde m es la multiplicidad
2. **Implementar el método de Newton**
3. **Analizar la convergencia para diferentes multiplicidades**
4. **Visualizar la función y el proceso iterativo**
5. **Comparar con el método de Newton modificado para raíces múltiples**


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton
````

````python

def funcion_con_raiz_multiple(x, multiplicidad=2, alpha=1.0):
    """Función con raíz múltiple en x=alpha"""
    return (x - alpha)**multiplicidad * np.cos(x - alpha)
````

````python
# grafique la función cooon raiz multiple 
x = np.linspace(-2, 4, 400)
y = funcion_con_raiz_multiple(x, multiplicidad=2, alpha=1.0)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = (x - 1)^2 * cos(x - 1)', color='blue')
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.axvline(1, color='red', linestyle='--', label='Raíz múltiple en x=1')
plt.title('Función con Raíz Múltiple')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
save_fig('funcion_con_raiz_multiple')
plt.show()
````

````python

def derivada_con_raiz_multiple(x, multiplicidad=2, alpha=1.0):
    """Derivada de la función con raíz múltiple"""
    return multiplicidad*(x-alpha)**(multiplicidad-1)*np.cos(x-alpha) - (x-alpha)**multiplicidad*np.sin(x-alpha)
````

````python

def metodo_newton(f, df, x0, tol=1e-10, max_iter=100):
    """Implementación básica del método de Newton"""
    iteraciones = [x0]
    for _ in range(max_iter):
        x_next = x0 - f(x0)/df(x0)
        iteraciones.append(x_next)
        if abs(x_next - x0) < tol:
            break
        x0 = x_next
    return np.array(iteraciones)
````

````python

def newton_modificado(f, df, multiplicidad, x0, tol=1e-10, max_iter=100):
    """Método de Newton modificado para raíces múltiples"""
    iteraciones = [x0]
    for _ in range(max_iter):
        x_next = x0 - multiplicidad * f(x0)/df(x0)
        iteraciones.append(x_next)
        if abs(x_next - x0) < tol:
            break
        x0 = x_next
    return np.array(iteraciones)
````

````python

def analizar_convergencia(multiplicidad=2):
    """Análisis completo para una multiplicidad dada"""
    alpha = 2.0
    f = lambda x: funcion_con_raiz_multiple(x, multiplicidad, alpha)
    df = lambda x: derivada_con_raiz_multiple(x, multiplicidad, alpha)
    
    # Punto inicial cerca de la raíz
    x0 = alpha + 0.5
    
    # Aplicar método de Newton estándar
    iter_newton = metodo_newton(f, df, x0)
    
    # Aplicar método de Newton modificado
    iter_modificado = newton_modificado(f, df, multiplicidad, x0)
    
    # Calcular errores
    errores_newton = np.abs(iter_newton - alpha)
    errores_modificado = np.abs(iter_modificado - alpha)
    
    # Graficar función y derivada
    plt.figure(figsize=(12, 10))
    
    # Gráfico de la función
    plt.subplot(2, 2, 1)
    x_vals = np.linspace(alpha-2, alpha+2, 400)
    plt.plot(x_vals, f(x_vals), label=f'f(x) = (x-{alpha})^{multiplicidad}*cos(x-{alpha})')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(alpha, color='red', linestyle='--', label=f'Raíz en x={alpha}')
    plt.title(f'Función con raíz de multiplicidad {multiplicidad}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    
    # Gráfico de la derivada
    plt.subplot(2, 2, 2)
    plt.plot(x_vals, df(x_vals), label=f"f'(x)")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(alpha, color='red', linestyle='--')
    plt.title('Derivada de la función')
    plt.xlabel('x')
    plt.ylabel("f'(x)")
    plt.legend()
    plt.grid(True)
    
    # Gráfico de convergencia
    plt.subplot(2, 2, 3)
    plt.semilogy(errores_newton, 'o-', label='Newton estándar')
    plt.semilogy(errores_modificado, 's-', label=f'Newton modificado (m={multiplicidad})')
    plt.title('Convergencia de los métodos')
    plt.xlabel('Iteración')
    plt.ylabel('Error absoluto (escala log)')
    plt.legend()
    plt.grid(True)
    
    # Gráfico de relación de convergencia
    plt.subplot(2, 2, 4)
    if len(errores_newton) > 1:
        ratios_newton = errores_newton[1:]/errores_newton[:-1]
        plt.plot(ratios_newton, 'o-', label='Ratio Newton')
    if len(errores_modificado) > 1:
        ratios_modificado = errores_modificado[1:]/errores_modificado[:-1]**2
        plt.plot(ratios_modificado, 's-', label='Ratio Newton modificado')
    plt.title('Ratio de convergencia')
    plt.xlabel('Iteración')
    plt.ylabel('Ratio de error')
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
    
    # Mostrar resultados numéricos
    print(f"\nAnálisis para multiplicidad = {multiplicidad}")
    print("Newton estándar:")
    print(f"Iteraciones: {len(iter_newton)}, Error final: {errores_newton[-1]:.2e}")
    print("\nNewton modificado:")
    print(f"Iteraciones: {len(iter_modificado)}, Error final: {errores_modificado[-1]:.2e}")
````

# Explicación del Ratio de Convergencia en el Análisis del Método de Newton



El ratio de convergencia es una medida fundamental para entender cómo de rápido un método iterativo se aproxima a la solución exacta. En el contexto del método de Newton para raíces múltiples, este concepto es especialmente importante para caracterizar el comportamiento del algoritmo.



## Concepto Matemático del Ratio de Convergencia



Dada una sucesión $\{x_{n}\}$ que converge a α, definimos el error en la iteración n como:
eₙ = |xₙ - α|

El **ratio de convergencia** en cada paso se calcula como:
Rₙ = eₙ₊₁ / eₙ^p

Donde p es el **orden de convergencia** que queremos analizar:
- p=1 para convergencia lineal
- p=2 para convergencia cuadrática



## Comportamiento en el Código Anterior

En el código proporcionado, calculamos dos tipos de ratios:

1. **Para el método de Newton estándar**:
   ```python
   ratios_newton = errores_newton[1:]/errores_newton[:-1]  # p=1
   ```
   Esto calcula Rₙ = eₙ₊₁/eₙ (convergencia lineal)

2. **Para el método de Newton modificado**:
   ```python
   ratios_modificado = errores_modificado[1:]/errores_modificado[:-1]**2  # p=2
   ```
   Esto calcula Rₙ = eₙ₊₁/eₙ² (convergencia cuadrática)

## Interpretación para Raíces Múltiples

Para una raíz de multiplicidad m:

1. **Método de Newton estándar**:
   - Orden de convergencia: 1 (lineal)
   - Ratio teórico: R → (m-1)/m cuando n → ∞
   - Ejemplo para m=2: R → 0.5

2. **Método de Newton modificado**:
   - Orden de convergencia: 2 (cuadrático)
   - Ratio teórico: R → constante (convergencia más rápida)

## Visualización en las Gráficas

En el cuarto subplot del código, graficamos estos ratios:
- Para Newton estándar (p=1): Esperamos que el ratio se estabilice alrededor de (m-1)/m
- Para Newton modificado (p=2): Esperamos que el ratio tienda a una constante (indicando convergencia cuadrática)

## Ejemplo Práctico

Para una raíz doble (m=2):
- Newton estándar: ratios ≈ 0.5, 0.5, 0.5,... (convergencia lineal)
- Newton modificado: ratios ≈ 0.3, 0.29, 0.3,... (pero con eₙ₊₁ ≈ C·eₙ²)

Esta diferencia explica por qué el método modificado converge mucho más rápido para raíces múltiples cuando conocemos la multiplicidad.

## Importancia en el Análisis

El cálculo del ratio nos permite:
1. Verificar experimentalmente el orden de convergencia teórico
2. Diagnosticar problemas cuando la convergencia no es la esperada
3. Comparar la eficiencia de diferentes métodos
4. Estimar cuántas iteraciones adicionales se necesitarán para alcanzar una precisión deseada

El análisis de ratios es particularmente valioso para métodos iterativos como Newton, donde el comportamiento de convergencia puede cambiar significativamente según las propiedades de la función (como la multiplicidad de las raíces).

## Solución  

$$ 2x^{2} - 4x + 3 = 2\left(x - \left(1+\frac{\sqrt{2}}{2} i \right) \right) \left(x - \left(1-\frac{\sqrt{2}}{2} i \right)\right) $$

## Conclusiones

1. Todos los trinomios de la forma $ax^{2} + bx + c$ factorizan en los complejos. (Y se hace con la f´´ormula del bachiller)
2. Algunos los trinomios de la forma $ax^{2} + bx + c$ factorizan en los Reales. (Es cuando el discriminante es mayor o igual a cero)
3. Algunos los trinomios de la forma $ax^{2} + bx + c$ factorizan en los Racionales (Es cuando el discriminante es un entero cuadrado perfecto).

 **Diseño de actividad de clase para fomentar trabajo en equipo, asertividad, empatía, inteligencia emocional, tolerancia, atención y puntualidad**.  





### **Actividad: "El Reto del Puente Emocional"**  
**Duración:** 15 minutos  
**Objetivos:**  
- Fomentar el **trabajo en equipo** y la **comunicación asertiva**.  
- Desarrollar **empatía** y **tolerancia** ante diferentes perspectivas.  
- Reducir el **temor a las pruebas** mediante técnicas de **inteligencia emocional**.  
- Promover la **atención** y **puntualidad** mediante dinámicas estructuradas.  




### **Estructura de la Actividad:**  

#### **1. Introducción (2 min)**  
- **Enfoque en puntualidad:**  
  - _"Hoy empezamos exactamente a tiempo, porque cada minuto cuenta. ¿Por qué creen que la puntualidad es clave para el trabajo en equipo?"_ (Breve reflexión grupal).  



#### **2. Dinámica de Roles y Empatía (5 min)**  
- **"Ponte en sus zapatos":**  
  - Se forman **tríos**: cada uno recibe un rol (Ej: **el impuntual**, **el que se enoja**, **el mediador**).  
  - Se les da una situación problemática (ej: "Un trabajo grupal se retrasa por la demora de un miembro").  
  - **Consigna:** Cada uno debe **argumentar desde su rol** y luego buscar una **solución asertiva**.  
  - **Reflexión final:** _"¿Cómo se sintieron en cada rol? ¿Qué estrategias usaron para resolver el conflicto?"_.  



#### **Cierre (1 min)**  
- **Compromiso grupal:**  
  - Cada estudiante dice **una palabra** sobre qué llevan de la actividad (ej: "empatía", "paciencia", "respirar antes de un examen").  
  - **Frase final:** _"Un equipo fuerte se construye con respeto, atención y puntualidad"_.  




### **Notas Adicionales:**  
- **Para reforzar la puntualidad:**  
  - Iniciar siempre a la hora exacta y felicitar a quienes lleguen temprano.  
- **Para la tolerancia:**  
  - Usar ejemplos cotidianos (ej: "¿Cómo reaccionarían si un compañero piensa distinto en un debate?").  
- **Para la atención:**  
  - Incluir cambios de tono de voz o movimientos físicos (levantar mano al escuchar una palabra clave).  

Esta actividad es **dinámica, vivencial y reflexiva**, ideal para adolescentes o adultos. ¡Puedes adaptarla según la edad del grupo!
