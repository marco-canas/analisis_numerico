
## 6. Función para Suma de Riemann (Precisión ε)


````python
def riemann_n(f, a, b, epsilon=0.0001):
    """
    Determina el número de rectángulos n necesario para alcanzar la precisión epsilon
    en la integral aproximada de f entre a y b usando Sumas de Riemann.
    
    Args:
        f (function): Función a integrar
        a, b (float): Límites de integración
        epsilon (float): Precisión deseada
    
    Returns:
        tuple: (n mínimo, error estimado)
    """
    n = 1
    error = float('inf')
    integral_prev = 0
    
    while error > epsilon:
        n *= 2  # Estrategia de duplicación
        x = np.linspace(a, b, n+1)
        delta = (b - a)/n
        integral = sum(f(a + i*delta) * delta for i in range(n))
        error = abs(integral - integral_prev)
        integral_prev = integral
        
    return n, error

# Ejemplo de uso:
def f(x): return x**2 + 2*x + 1
n, error = riemann_n(f, 0, 5)
print(f"Se requieren {n} rectángulos para error < 0.0001")

````

El resultado de observaciones de un proceso de laboratorio o industrial genera una lista de datos que forman tablas.

Por ejemplo, la medición de la densidad del carbonato de sodio a distintas concentraciones a temperatura constante genera los siguientes datos.    


Los datos expresan la relación de la densidad (variable dependiente) y la concentración (variable independiente). 

Es decir que el valor de y depende del valor de x. Si deseamos conocer el valor de la variable dependiente para un valor de la variable independiente que no está en la tabla de datos, se requiere hacer una estimación.

La estimación puede lograrse de dos maneras, ajustando una función que toque todos los puntos de la tabla (interpolación) o ajustando una función que minimice las distancias del punto tabulado a la función (regresión).

La función de la interpolación pasa por cada punto de la tabla, la función de la regresión no necesariamente pasa lo más cerca de los puntos. 

Esto es, si sustituimos $x_{i}$ en $f(x)$, el resultado es $y_{i}$ para la interpolación y para la regresión obtenemos $y_{i}$ corregida.

````python
import numpy as np
from scipy import interpolate
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Datos experimentales
x = np.array([4.1, 12.2, 20.3, 28.2, 38.1, 45.2])
y = np.array([1.0276, 1.1013, 1.1801, 1.2652, 1.3480, 1.4120])

# Puntos finos para graficar las curvas suaves
xfino = np.linspace(min(x), max(x), 100)

# 1. Interpolación lineal
interpolador = interpolate.interp1d(x, y, kind='linear')
yi = interpolador(xfino)  # Valores interpolados

# 2. Regresión lineal
def modelo_lineal(x, a, b):
    return a + b * x

# Ajuste por mínimos cuadrados
parametros, _ = curve_fit(modelo_lineal, x, y)
a, b = parametros
yr = modelo_lineal(xfino, a, b)  # Valores de regresión

# Gráfico comparativo
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'ko', markersize=8, label='Datos experimentales')
plt.plot(xfino, yi, 'b-', linewidth=2, label='Interpolación lineal')
plt.plot(xfino, yr, 'r--', linewidth=2, label='Regresión lineal')

# Configuración del gráfico
plt.title('Comparación: Interpolación Lineal vs. Regresión Lineal', fontsize=14)
plt.xlabel('Variable independiente (x)', fontsize=12)
plt.ylabel('Variable dependiente (y)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar ecuación de regresión
texto_regresion = f'Regresión: y = {a:.4f} + {b:.4f}x'
plt.text(0.05, 0.9, texto_regresion, transform=plt.gca().transAxes, 
         fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
````

## 2.1 Interpolación   

El objetivo es ajustar una función que por combinaciones lineales logre pasar por todos los puntos.  

### Definición 2.1.1 — Interpolación. 

Las técnicas de interpolación ajustan un polinomio de menor grado que pasa por los puntos tabulados,esto quiere decir que, si sustituimos un valor de xi de la tabla, obtenemos la correspondiente yi de la tabla. En general, la función tiene la forma:

$$ f(x) = a_{0}g_{0}(x) + a_{1}g_{1}(x) + a_{2}g_{2}(x) + a_{3}g_{3}(x) + \cdots + a_{n}g_{n}(x)  $$

Donde las $a_{i}$ son factores a determinar y las gi(x) son funciones de una familia en particular.

## Familia de monomios.   

Son los más utilizados ya que son fáciles de obtener y generan un polinomio. La desventaja es que los polinomios tienen comportamientos diferentes a los esperados, de acuerdo con los datos, cuando son de potencias altas:  

$$ f(x) = a_{0} + a_{1}x + a_{2}x^{2} + a_{3}x^{3} + \cdots + a_{n}x^{n} $$

$$ f(x) = \sum_{i = 0}^{n} a_{i}x^{i} $$

La técnica de ajustar un polinomio a los datos es la más usada porque se adapta a muchos casos simples, las otras técnicas se usan para un conjunto de datos que tienen un comportamiento cíclico o exponencial.

### 2.1.1 Interpolación lineal  

La interpolación lineal ajusta un polinomio de grado 1 a los datos tabulados. Usamos entonces la forma de una ecuación lineal

$$ f(x) = a + bx $$

Para lograr una interpolación lineal (polinomio de grado 1), se requieren dos puntos de la tabla  

|x|y|
|-|-|
|x_{0}|y_{0}|
|x_{1}|y_{1}|

Como indica la definición 2.1.1 se debe encontrar un polinomio que pase por los puntos tabulados. Entonces, si sustituimos x0 en f (x) (Ecuación lineal 2.2 ) debemos obtener y0.

$$ f(x_{0}) = a + bx_{0} = y_{0} $$
Si sustituimos x1 en f (x) (Ecuación lineal 2.2 ) debemos obtener y1.
$$ f(x_{1}) = a + bx_{1} = y_{1} $$

Tenemos dos ecuaciones (2.3 y 2.4) y dos incógnitas (a y b), entonces procedemos a resolver el sistema de 2x2.

Despejamos a de ecuación 2.3 

$$ a = y_{0} - bx_{0} $$

Sustituimos a en la ecuación 2.4
$$ y_{1} = (y_{0} - bx_{0}) + bx_{1} $$

$$ b = \frac{y_{1}- y_{0}}{x_{1} - x_{0}} $$

Finalmente obtenemos el modelo de interpolación lineal:

$$ f(x) = y_{0} + \frac{y_{1}- y_{0}}{x_{1} - x_{0}}(x - x_{0}) $$

# 2.1.3 Polinomio de interpolación de Newton     

Como observamos en la interpolación lineal (ecuación 2.2) y la interpolación cuadrática (ecuación 2.9)   cada término que se va aumentando es una diferencia dividida de orden mayor.  

El polinomio de interpolación polinómica para $n$ datos tiene la forma:  

$$ f(x) = a_{0} + a_{1}(x - x_{0}) + a_{2}(x-x_{0})(x-x_{1}) + \cdots + a_{n}(x - x_{0})(x - x_{1}) \cdots (x - x_{n-1}) $$

donde  

$a_{0} = y_{0}$  

$a_{1} = \frac{y_{1} - y_{0}}{x_{1} - x_{0}} = f[x_{1}, x_{0}]$

$a_{2} = \frac{\frac{y_{2} - y_{1}}{x_{2} - x_{1}} - \frac{y_{1} - y_{0}}{x_{1} - x_{0}}}{x_{2} - x_{0}} = \frac{f[x_{2}, x_{1}] - f[x_{1}, x_{0}]}{x_{2} - x_{0}} = f[x_{2}, x_{1}, x_{0}]$

$a_{n} = f[x_{n}, x_{n-1}, \ldots, x_{0}]$

La **fórmula de interpolación de Newton** en términos de **diferencias divididas** se expresa en LaTeX de la siguiente manera:



### **Fórmula General**:



$$ P_n(x) = f[x_0] + \sum_{k=1}^{n} \left( f[x_0, x_1, \dots, x_k] \cdot \prod_{j=0}^{k-1} (x - x_j) \right) $$




### **Explicación de los componentes**:
1. **Diferencias divididas**:
   - **Primer orden**: 

$$ f[x_i, x_{i+1}] = \frac{f(x_{i+1}) - f(x_i)}{x_{i+1} - x_i} $$



   - **Orden superior** (recursivo): 

 
$$ f[x_i, \dots, x_{i+k}] = \frac{f[x_{i+1}, \dots, x_{i+k}] - f[x_i, \dots, x_{i+k-1}]}{x_{i+k} - x_i} $$
     



2. **Términos del polinomio**:
   - El producto $\prod_{j=0}^{k-1} (x - x_j)$ representa los factores $(x - x_0)(x - x_1)\dots(x - x_{k-1})$.



### **Ejemplo completo en LaTeX**:




El polinomio de interpolación de Newton con diferencias divididas es:

$$
P_n(x) = f[x_0] + f[x_0, x_1](x - x_0) + f[x_0, x_1, x_2](x - x_0)(x - x_1) + \dots + f[x_0, \dots, x_n]\prod_{j=0}^{n-1} (x - x_j)
$$

donde las diferencias divididas se calculan recursivamente:

$$
f[x_i, \dots, x_{i+k}] = \frac{f[x_{i+1}, \dots, x_{i+k}] - f[x_i, \dots, x_{i+k-1}]}{x_{i+k} - x_i}
$$






### **Notas**:
- **Ventaja de la forma de Newton**: Permite añadir nuevos puntos de interpolación sin recalcular todo el polinomio (solo añadiendo un término más).
- **Uso común**: En algoritmos numéricos para interpolación o construcción de tablas de diferencias divididas.

# Solución completa con generación de datos climáticos simulados para Caucasia-Antioquia


````python

# Configuración inicial para notebook interactivo
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import ipywidgets as widgets
from scipy.interpolate import interp1d

%matplotlib inline

# Generación de datos climáticos simulados para Caucasia-Antioquia (promedios mensuales 2010-2020)
np.random.seed(42)  # Para reproducibilidad

# Meses del año (1 a 12)
meses = np.arange(1, 13)

# Temperaturas promedio mensuales en °C (simuladas con patrón estacional)
temp_base = 28.5  # Temperatura base anual
amplitud = 1.5    # Amplitud de variación estacional
ruido = np.random.normal(0, 0.3, size=len(meses))  # Ruido aleatorio

# Modelo de temperatura con patrón sinusoidal + ruido
temps = temp_base + amplitud * np.sin(2 * np.pi * (meses - 4) / 12) + ruido

# Ajustamos algunos valores para mayor realismo
temps[3] += 0.7  # Abril más caliente
temps[7] -= 0.5  # Agosto más fresco
temps = np.round(temps, 1)  # Redondeamos a 1 decimal

print("Datos climáticos generados:")
print("Meses:", meses)
print("Temperaturas (°C):", temps)

# Ejemplo de animación comparativa
fig, ax = plt.subplots(figsize=(10, 6))

def update(frame):
    ax.clear()
    grado = frame + 1
    coeffs = np.polyfit(meses, temps, grado)
    modelo = np.poly1d(coeffs)
    
    ax.scatter(meses, temps, color='red', label='Datos reales')
    x_smooth = np.linspace(1, 13, 100)
    ax.plot(x_smooth, modelo(x_smooth), label=f'Regresión (grado {grado})')
    
    # Configuración del gráfico
    ax.set_title('Modelado de Temperaturas en Caucasia\nEfecto del Grado en Regresión Polinomial')
    ax.set_xlabel('Mes del año')
    ax.set_ylabel('Temperatura (°C)')
    ax.set_xticks(meses)
    ax.set_xticklabels(['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 
                       'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend()
    
    # Añadir información adicional
    r_squared = 1 - sum((temps - modelo(meses))**2)/sum((temps - np.mean(temps))**2)
    ax.text(0.02, 0.95, f'R² = {r_squared:.3f}', transform=ax.transAxes,
            bbox=dict(facecolor='white', alpha=0.8))

# Crear la animación
ani = FuncAnimation(fig, update, frames=5, interval=1500)

# Mostrar la animación
HTML(ani.to_jshtml())

````
