**Ejemplo 2 ‚Äì Producci√≥n agr√≠cola con ley de rendimientos decrecientes**

## üîπ 2. Contextos **F√≠sicos (Cinem√°tica)**

Las ecuaciones de movimiento pueden ser no lineales.

**Ejemplo 1 ‚Äì Movimiento vertical con resistencia cuadr√°tica**

## üîπ 3. Contextos **Econ√≥micos**

Las ecuaciones de equilibrio de mercados o de inversi√≥n suelen ser no lineales.

**Ejemplo 1 ‚Äì Tasa Interna de Retorno (TIR)**
La TIR se obtiene resolviendo:

$$
f(r) = \sum_{t=0}^n \frac{C_t}{(1+r)^t} = 0
$$

donde $C_t$ son los flujos de caja.
Como no hay f√≥rmula cerrada para $r$, se aplica Newton‚ÄìRaphson.

**Ejemplo 2 ‚Äì Modelo de oferta y demanda no lineal**
Supongamos:

$$
Q_d(p) = a - b \ln(p), \quad Q_s(p) = c + dp^2
$$

El equilibrio se obtiene resolviendo:

$$
f(p) = a - b \ln(p) - (c + dp^2) = 0
$$

Newton‚ÄìRaphson permite encontrar el precio de equilibrio $p^*$.
