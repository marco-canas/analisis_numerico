# **prototipo de cuaderno Jupyter** en Python para ver cómo se aplica el **método de Newton–Raphson** en un problema sencillo de optimización vinculado al entrenamiento de una red neuronal muy básica.

La idea es que vean:

1. La implementación del método Newton–Raphson.
2. Cómo se aplica para encontrar mínimos de una función de pérdida (ejemplo: función cuadrática).
3. Cómo se conecta esto con la optimización en redes neuronales (entrenamiento de un perceptrón simple).

Aquí tienes el prototipo:

---


```python
# ==========================================
# Proyecto: Método de Newton-Raphson aplicado a Deep Learning
# Prototipo de implementación en Jupyter Notebook
# ==========================================

import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 1. Implementación genérica de Newton-Raphson
# -----------------------------
def newton_raphson(f, f_prime, f_double_prime, x0, tol=1e-6, max_iter=50):
    """
    Método de Newton-Raphson para encontrar raíces o puntos críticos.

    Parámetros:
    f  : función objetivo
    f_prime : primera derivada
    f_double_prime : segunda derivada
    x0 : valor inicial
    tol : tolerancia
    max_iter : número máximo de iteraciones
    """
    x = x0
    history = [x0]
    
    for i in range(max_iter):
        fx = f_prime(x)
        dfx = f_double_prime(x)
        
        if abs(dfx) < 1e-8:  # evitar divisiones por cero
            print("Derivada segunda demasiado pequeña, detener iteración.")
            break
        
        x_new = x - fx / dfx
        history.append(x_new)
        
        if abs(x_new - x) < tol:
            break
        x = x_new
    
    return x, history

# -----------------------------
# 2. Ejemplo sencillo: minimizar una función cuadrática
# -----------------------------
# f(x) = (x-3)^2 + 2
f = lambda x: (x-3)**2 + 2
f_prime = lambda x: 2*(x-3)
f_double_prime = lambda x: 2

x0 = 0.0
optimo, hist = newton_raphson(f, f_prime, f_double_prime, x0)

print("Mínimo encontrado en x =", optimo)
print("Valor mínimo f(x) =", f(optimo))
```

```python

# Graficar la función y el proceso iterativo
x_vals = np.linspace(-1, 6, 200)
y_vals = f(x_vals)

plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label="f(x)")
plt.scatter(hist, [f(x) for x in hist], color="red", zorder=5, label="Iteraciones NR")
plt.title("Newton-Raphson aplicado a una función cuadrática")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()

# -----------------------------
# 3. Aplicación básica en un perceptrón simple
# -----------------------------
# Queremos minimizar la función de pérdida MSE para un perceptrón con un solo peso
X = np.array([1, 2, 3, 4], dtype=float)  # entradas
y = np.array([2, 4, 6, 8], dtype=float)  # salidas (y = 2x idealmente)

# Función de pérdida: MSE en términos del peso w
def loss(w):
    y_pred = w * X
    return np.mean((y - y_pred)**2)

# Primera derivada (gradiente respecto a w)
def loss_prime(w):
    y_pred = w * X
    return -2 * np.mean(X * (y - y_pred))

# Segunda derivada (Hessiano respecto a w)
def loss_double_prime(w):
    return 2 * np.mean(X**2)

# Aplicar Newton-Raphson al problema de entrenamiento
w0 = 0.5
w_opt, w_hist = newton_raphson(loss, loss_prime, loss_double_prime, w0)

print("\nPeso óptimo encontrado:", w_opt)
print("Error cuadrático medio con w*:", loss(w_opt))

# Graficar convergencia de la pérdida
w_vals = np.linspace(0, 3, 200)
loss_vals = [loss(w) for w in w_vals]

plt.figure(figsize=(8,5))
plt.plot(w_vals, loss_vals, label="Función de pérdida (MSE)")
plt.scatter(w_hist, [loss(w) for w in w_hist], color="red", label="Iteraciones NR")
plt.title("Optimización del peso de un perceptrón simple con Newton-Raphson")
plt.xlabel("Peso w")
plt.ylabel("MSE")
plt.legend()
plt.grid()
plt.show()

```

## 1. Introducción teórica

# Método de Newton–Raphson en Deep Learning

El método de Newton–Raphson es un algoritmo iterativo que se usa para encontrar raíces o puntos críticos de funciones. 
En optimización, lo aplicamos para encontrar **mínimos o máximos** de funciones de costo.

En Deep Learning, los algoritmos de entrenamiento de redes neuronales están basados en optimización, 
normalmente con **gradiente descendente**. Sin embargo, el método de Newton–Raphson introduce 
una idea más potente: usar tanto la **primera derivada (gradiente)** como la **segunda derivada (Hessiano)** 
para mejorar la convergencia.



## Fórmula general
Dado un parámetro $w$, actualizamos con:

$$ w_{k+1} = w_k - \frac{f'(w_k)}{f''(w_k)} $$

Donde:
- $f'(w)$ es la primera derivada (gradiente de la pérdida).
- $f''(w)$ es la segunda derivada (curvatura de la pérdida).




## 2. Implementación general en Python



```python

import numpy as np
import matplotlib.pyplot as plt

def newton_raphson(f_prime, f_double_prime, w0, tol=1e-6, max_iter=50):
    """
    Método Newton-Raphson para optimización unidimensional.
    """
    w = w0
    history = [w0]
    
    for i in range(max_iter):
        grad = f_prime(w)
        hess = f_double_prime(w)
        
        if abs(hess) < 1e-8:  # evitar división por cero
            print("Hessiano demasiado pequeño, detener iteración.")
            break
        
        w_new = w - grad / hess
        history.append(w_new)
        
        if abs(w_new - w) < tol:
            break
        w = w_new
    
    return w, history


```



## 3. Ejemplo con una función no cuadrática (pérdida logística)


# Ejemplo 1: Función de pérdida logística
Consideremos un problema de clasificación binaria simple con función de pérdida logística:

$$ L(w) = \log(1 + e^{-yw}) $$

donde:
- $w$ es el parámetro a optimizar.
- $y \in \{-1,1\}$ es la etiqueta.

Este tipo de pérdida es más compleja que la cuadrática y aparece en regresión logística 
y redes neuronales con salida sigmoidal.



```python
# Definimos la pérdida logística, su gradiente y su hessiano
y = 1  # etiqueta positiva

loss = lambda w: np.log(1 + np.exp(-y*w))
loss_prime = lambda w: -y * np.exp(-y*w) / (1 + np.exp(-y*w))
loss_double_prime = lambda w: (np.exp(-y*w)) / ((1 + np.exp(-y*w))**2)

# Aplicar Newton-Raphson
w0 = -2.0
w_opt, hist = newton_raphson(loss_prime, loss_double_prime, w0)

print("Parámetro óptimo:", w_opt)

# Visualización
w_vals = np.linspace(-5, 5, 200)
loss_vals = [loss(w) for w in w_vals]

plt.figure(figsize=(8,5))
plt.plot(w_vals, loss_vals, label="Función de pérdida logística")
plt.scatter(hist, [loss(w) for w in hist], color="red", label="Iteraciones NR")
plt.title("Newton-Raphson aplicado a la pérdida logística")
plt.xlabel("w")
plt.ylabel("L(w)")
plt.legend()
plt.grid()
plt.savefig(r'C:\Users\marco\Downloads\funcion_perdida_logistica.png')
plt.show()

```


## 4. Ejemplo aplicado: un perceptrón con activación sigmoide




# Ejemplo 2: Entrenamiento de un perceptrón con activación sigmoide

Consideremos un perceptrón simple con función sigmoide:

$$ \hat{y} = \sigma(wx) = \frac{1}{1 + e^{-wx}} $$

La función de pérdida será el **log-loss** (entropía cruzada binaria):

$$ L(w) = - \frac{1}{n} \sum_{i=1}^n [ y_i \log(\hat{y}_i) + (1-y_i)\log(1-\hat{y}_i) ] $$

Este caso es más realista que la función cuadrática, y se conecta directamente con 
redes neuronales en problemas de clasificación.



```python
# Datos simulados para clasificación binaria
X = np.array([0.5, 1.0, 2.0, 3.0])
y = np.array([0, 0, 1, 1])  # etiquetas binarias

# Definición de funciones
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def loss(w):
    y_pred = sigmoid(w * X)
    return -np.mean(y * np.log(y_pred + 1e-8) + (1-y)*np.log(1 - y_pred + 1e-8))

def loss_prime(w):
    y_pred = sigmoid(w * X)
    return np.mean((y_pred - y) * X)

def loss_double_prime(w):
    y_pred = sigmoid(w * X)
    return np.mean(X**2 * y_pred * (1 - y_pred))

# Newton-Raphson aplicado a este modelo
w0 = 0.1
w_opt, hist = newton_raphson(loss_prime, loss_double_prime, w0)

print("Peso óptimo encontrado:", w_opt)
print("Pérdida en el óptimo:", loss(w_opt))
```

```python

# Graficar pérdida
w_vals = np.linspace(-5, 5, 200)
loss_vals = [loss(w) for w in w_vals]

plt.figure(figsize=(8,5))
plt.plot(w_vals, loss_vals, label="Función de pérdida (log-loss)")
plt.scatter(hist, [loss(w) for w in hist], color="red", label="Iteraciones NR")
plt.title("Optimización de un perceptrón logístico con Newton-Raphson")
plt.xlabel("w")
plt.ylabel("L(w)")
plt.legend()
plt.grid()
plt.savefig(r"C:\Users\marco\Downloads\optimizacion_w.png")
plt.show()
```

# Reflexión final

- El método de Newton–Raphson converge más rápido que el gradiente descendente 
  en muchos casos, porque usa información de la curvatura (segunda derivada).
  
- Sin embargo, en redes neuronales con **muchos parámetros** el cálculo del Hessiano 
  es costoso, y por eso en Deep Learning moderno se usan métodos aproximados 
  como **Quasi-Newton (BFGS, L-BFGS)** o algoritmos de primer orden más eficientes 
  como **Adam**.
