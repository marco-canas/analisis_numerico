## **Contexto del problema**

En el tiro parab√≥lico con resistencia del aire proporcional a la velocidad, las ecuaciones de movimiento en el eje vertical se modelan como:

$$
v(t) = v_0 e^{-\frac{k}{m}t} - \frac{mg}{k}\big(1 - e^{-\frac{k}{m}t}\big)
$$

$$
h(t) = \frac{m}{k}\left(v_0 + \frac{mg}{k}\right)\left(1 - e^{-\frac{k}{m}t}\right) - \frac{mg}{k}t
$$

donde:

* $v(t)$ es la velocidad en funci√≥n del tiempo.
* $h(t)$ es la altura en funci√≥n del tiempo.
* $v_0$ = velocidad inicial (m/s).
* $m$ = masa del proyectil (kg).
* $k$ = coeficiente de rozamiento del aire (kg/s).
* $g$ = 9.8 m/s$^2$.
