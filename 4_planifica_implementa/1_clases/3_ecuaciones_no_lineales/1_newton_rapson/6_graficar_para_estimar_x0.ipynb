# La importancia de graficar para estimar la aproximación inicial al cero de una función  

Meta: estimar los tres ceros de la siguiente función polinómica..  

$$ 
f(x) = x^{3} - 2x + 1
$$

# Solución: 



```python
import matplotlib.pyplot as plt # visualizar datos
import numpy as np  # crear el dominio y rango de la función

```

```python
dominio = np.linspace(-1.8, 1.2, 100)  # dominio de la función
rango = dominio**3 - 2*dominio + 1  # rango de la función

plt.axhline(0, color='black', lw=0.5)  # eje x
plt.axvline(0, color='black', lw=0.5)  # eje y
plt.plot(dominio, rango, label='$f(x) = x^{3} - 2x + 1$')
plt.title('Estimación gráfica de los ceros de la función')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend()
plt.savefig(r"C:\Users\marco\Downloads\estimacion_grafica_ceros.png")
plt.show()
```

```python
x0_primero = -1.5
x0_segundo = 0.5
x0_tercero = 1.0
```

```python
def f(x): 
    return x**3 - 2*x + 1

def f_prima(x):
    return 3*x**2 - 2
```

```python

for i in range(1, 5):
    x1_primero = x0_primero - f(x0_primero)/f_prima(x0_primero)
    x0_primero = x1_primero

x0_primero
```

```python
f(x0_primero)
```

```python
for i in range(1, 6):
    x1_segundo = x0_segundo - f(x0_segundo)/f_prima(x0_segundo)
    x0_segundo = x1_segundo

x0_segundo
```

```python
f(x0_segundo)
```

```python
for i in range(1, 6):
    x1_tercero = x0_tercero - f(x0_tercero)/f_prima(x0_tercero)
    x0_tercero = x1_tercero

x0_tercero
```

```python
f(x0_tercero)
```

Conclusión hemos encontrado los tres ceros de la función: y son aproximadamente: 
- Primer cero: -1.618
- Segundo cero: 0.618
- Tercer cero: 1.0
