```python
import numpy as np 
from scipy import stats 
import matplotlib.pyplot as plt 
```

```python
# Aqu√≠ estamos ingresando los datos 
y = np.array([6,9,11,13,22,26,28,33,35])
x = np.array([68,67,65,53,44,40,37,34,32])
```

```python
m, b, r, p, std_err = stats.linregress(x, y)
```

```python
m 
```

```python
b
```

```python
r
```

```python
print('Intersecto con el eje y ', b)
print('pendiente ', m)
print('coeficiente de correlacion ', r)
```

```python
 y_pred = m*x+b
```

```python
y_pred 
```

```python
plt.plot(x,y,'o', label='Datos')
plt.plot(x,y_pred,'r:',label='Regresion lineal')

plt.title('Deformacion y Dureza')
plt.xlabel('Dureza')
plt.ylabel('Deformacion')
```
