## Ejercicio 1: Variables y tipos

```python
entero = 10
decimal = 3.14
cadena = "Hola"
lista = [1,2,3]
diccionario = {"a":1, "b":2}

variables = [entero, decimal, cadena, lista, diccionario]
for v in variables:
    print(v, type(v))
```

## Ejercicio 2: Suma de naturales

```python
n = 10
suma = 0
for i in range(1, n+1):
    suma += i
suma
```

## Ejercicio 3: Contar vocales

```python
texto = "Python es excelente"
vocales = "aeiouáéíóúAEIOUÁÉÍÓÚ"
contador = 0
for c in texto:
    if c in vocales:
        contador += 1
contador
```

## Ejercicio 4: Función cuadrado

```python
def cuadrado(x):
    return x*x

for i in range(1,11):
    print(i, cuadrado(i))
```

## Ejercicio 5: Función es_primo

```python
def es_primo(n):
    if n<2: return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True

primos = [i for i in range(1,51) if es_primo(i)]
primos
```
